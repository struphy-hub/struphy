# MHD-drift-kinetic hybrid models

This is a collection of hybrid models we will consider during the thesis. The notebook includes:

1. Geometric structures (Lagrangians, Poisson brackets) on the continuous level.
2. Dispersion relations.
3. 2D and 3D de Rham complexes and corresponding construction of basis functions (splines, Lagrange).

References are:

* _Stefan's notes_ on hybrid modeling in current-coupling scheme (CCS).
* J. Burby, C. Tronci, _Variational approach to low-frequency kinetic-MHD in the current coupling scheme_, PPCF 2017.


# 1. Models and dispersion relations

## Standard MHD

In standard MHD the electric field takes the simple form

$$
\be \label{sub1:E}
 \Eb = -\Ub \times \Bb\,,
\ee
$$

where $\Ub$ stands for the plasma bulk velocity. The standard, compressible MHD equations read

$$
\begin{align}
 &\pder{\tau}{t} + \nabla \cdot (\tau \Ub) = 0\,,
 \\[2mm]
 &\pder{\Ub}{t} + \Ub \cdot \nabla \Ub + \frac{\nabla p}{\tau}
  =  \frac{1}{\mu_0}\frac{\nabla \times 
\Bb }{\tau} \times\Bb   \,,
 \\[2mm]
 &\pder{p}{t} +  \Ub \cdot \nabla p + \gamma\, p\, \nabla \cdot \Ub  = 0\,, 
 \\[2mm]
 & \pder{\Bb}{t} = \nabla \times (\Ub \times \Bb)\,.
\end{align}
$$

Here, $\tau$ denotes the plasma mass, $p$ is the plasma pressure and $\gamma = 5/3$ is the specific heat ratio of an ideal gas in 3D.

### Geometric structure

### Linear MHD and dispersion relation

Linearized MHD equations are obtained by keeping only the first 
order terms in an expansion of the solution around the homogeneous equilibrium

$$
\be \label{equil:hom}
 \tau_0 = const. \,,\qquad \Ub_0 = 0\,,\qquad p_0 = const.\,,\qquad \Bb_0 = B_0\,\eb_z\,.
\ee
$$

Writing $\tau = \tau_0 + \tau_1$ and relabeling $\tau_1 \to \tau$ after linearization 
yields the system

$$
\begin{align}
 &\pder{\tau}{t} + \tau_0\nabla \cdot \Ub = 0\,,
 \\[2mm]
 &\pder{\Ub}{t} + \frac{\nabla p}{\tau_0} =  \frac{B_0}{\tau_0\mu_0}\, (\nabla \times 
\Bb ) \times \eb_z   \,,  \label{lin1:U}
 \\[2mm]
 &\pder{p}{t} + \gamma\, p_0\, \nabla \cdot \Ub  = 0\,, 
 \\[2mm]
 & \pder{\Bb}{t} = B_0\,\nabla \times (\Ub \times \eb_z)\,.
\end{align}
$$

We are interested in waves propagating in the direction of the magnetic field, hence with wave 
vector $\kb = k_\para\eb_z$. Writing the unknowns as $\wb = (\tau,\Ub,p,\Bb)^\top \in \RR^8$, we 
insert the ansatz $\wb(z,t) = \wh \wb\, e^{i (k_\para z - \omega t)}$ with $\wh 
\wb \in\RR^8$ constant to obtain the eigenvalue problem

$$
\be \label{eig1}
 - \omega\,\wb + \Ab \cdot \wb = 0\,,
\ee
$$

where the matrix $\Ab \in \RR^{8\times 8}$ can be deduced from

$$
\begin{align}
 &-\omega\,\tau + \tau_0\, k_\para U_\para = 0\,,
 \\[2mm]
 &-\omega\,\Ub + \frac{1}{\tau_0} k_\para\, p\,\eb_z - \frac{B_0}{\tau_0\mu_0}\,k_\para\Bb_\perp = 
0   \,,
 \\[2mm]
 &- \omega\,p + \gamma\, p_0\, k_\para U_\para  = 0\,, 
 \\[3mm]
 & -\omega\,\Bb - B_0\,k_\para\Ub_\perp = 0 \,,
\end{align}
$$

and we used the notation $U_\para = \Ub \cdot \eb_z$ and $\Ub_\perp = \eb_z \times \Ub \times 
\eb_z = (U_x,U_y,0)^\top$. Solving \eqref{eig1} is straightforward: from the above we obtain

$$
\be \label{lin1sol:a}
 \tau = \frac{\tau_0}{\omega}\, k_\para U_\para \,,\qquad  p = \frac{\gamma\, p_0}{\omega}\, 
k_\para U_\para  \,,\qquad \Bb = \Bb_\perp = - \frac{B_0}{\omega}\,k_\para\Ub_\perp \,,
\ee
$$

which inserted into the momentum equation leads to

$$
\be \label{lin1sol:b}
 \omega^2\,\Ub - \frac{\gamma\,p_0}{\tau_0}\, k_\para^2\, U_\para\,\eb_z - 
\frac{B_0^2}{\tau_0\mu_0}\,k_\para^2\,\Ub_\perp = 0\,.
\ee
$$

The two relevant velocities are the sound speed $v_s$ and the Alfv\'en velocity $v_A$,

$$
\be
 v_\tn{s} = \sqrt{\frac{\gamma\,p_0}{\tau_0}}\,,\qquad v_\tn{A} = 
\sqrt{\frac{B_0^2}{\tau_0\mu_0}}\,,
\ee
$$

such that \eqref{lin1sol:b} reads

$$
\be
 \begin{pmatrix}
  \omega^2 - v_\tn{A}^2\,k_\para^2 & 0 & 0 \\
  0 & \omega^2 - v_\tn{A}^2\,k_\para^2 & 0 \\
  0 & 0 & \omega^2 - v_\tn{s}^2 \,k_\para^2
 \end{pmatrix}
 \begin{pmatrix}
  U_x \\ U_y \\ U_\para
 \end{pmatrix}
 = 0\,.
\ee
$$

Hence, there are two kinds of eigenmodes:

1. Sound waves: $\omega = \pm c_\tn{s} k_\para$. In this case $\Ub_\perp = 0$ and thus there 
is no perturbation of the magnetic field due to \eqref{lin1sol:a}. 

2. Shear Alfv\'en waves: $\omega = \pm v_\tn{A} k_\para$. In this case $U_\para = 0$, the 
magnetic wave is transversal and there is no perturbation of the density and the 
pressure due to \eqref{lin1sol:a}. 

## MHD-drift-kinetic current coupling

Here, standard MHD is coupled to the drift kinetic equation for the hot ions [Tronci, Burby 2017]:

$$
\begin{align}
 &\pder{\tau}{t} + \nabla \cdot (\tau \Ub) = 0\,,
 \\[3mm]
 &\pder{\Ub}{t} + \Ub \cdot \nabla \Ub + \frac{\nabla p}{\tau}
  =   \frac{1}{\mu_0}\frac{\nabla \times 
\Bb }{\tau} \times\Bb + \frac{(q_\tn{h}\nhD\,\Ub - \jhD )}{\tau} \times\Bb   
\,,
 \\[3mm]
 &\pder{p}{t} +  \Ub \cdot \nabla p + \gamma\, p\, \nabla \cdot \Ub  = 0\,, 
 \\[3mm]
 & \pder{\Bb}{t} = \nabla \times (\Ub \times \Bb)\,,
 \\[3mm]
 & \pder{\fhD}{t} + \ub_\tn{g} \cdot \nabla \fhD + E_\tn{g}\,
\pder{\fhD}{v_\para} = 0\,,
\end{align}
$$

where $\fhD = \fhD(t,\rb,v_\para,\wh\mu)$ is the guiding-center (GC) distribution, $\rb$ denotes the GC position, $v_\para$ the parallel GC velocity and $\wh\mu$ is a constant of the motion (generalized magnetic moment). Moreover,

$$
\be \label{sub3:ugEg}
 \ub_\tn{g} =  v_\para\,\frac{\Bb^*}{B^*_\para} + 
\frac{\Eb^* \times 
\bb}{B^*_\para} \,,\qquad E_\tn{g} = \frac{q_\tn{h}}{m_\tn{h}} \frac{\Eb^* \cdot 
\Bb^*}{B^*_\para} \,,
\ee
$$

where $q_\tn{h}$ and $m_\tn{h}$ stand for the hot particle's charge and mass, respectively, and

$$
\begin{align}
 \Bb^* &= \Bb + \frac{m_\tn{h}}{q_\tn{h}} v_\para\nabla \times \bb\,,
 \\[2mm]
 \Eb^* &= -\Ub \times \Bb - \frac{m_\tn{h}}{q_\tn{h}} \wh \mu \nabla B - \frac{m_\tn{h}}{q_\tn{h}} v_\para  \pder{\bb}{t}\,,
 \\[2mm]
 \nhD &= \int \fhD\,B^*_\para\,\tn dv_\para\tn d\wh\mu\,,
 \\[2mm]
 \jhD &= q_\tn{h} 
\int_{\wh\mu}\ub_\tn{g}\, 
\fhD\,B^*_\para\,\tn dv_\para + \nabla \times \Mb\,,
 \\[2mm]
 \Mb &= - m_\tn{h}\int\bigg[ \wh\mu\,\bb -
\frac{v_\para}{B}\,\ub_{\tn{g},\perp} \bigg]\,\fhD\,B^*_\para\,\tn 
dv_\para \tn d\wh\mu\,,  \label{sub3:M}
\end{align}
$$

with $B^*_\para = \Bb^* \cdot \bb$.

### Geometric structure of the full model 

### Linear MHD part  

Let us consider a coupling of the linear MHD equations (from section 1.1) with the full drift-kinetic equation:

$$
\begin{align}
 &\pder{\tau}{t} + \tau_0\nabla \cdot \Ub = 0\,,
 \\[2mm]
 &\pder{\Ub}{t} + \frac{\nabla p}{\tau_0} =  \frac{B_0}{\tau_0\mu_0}\, (\nabla \times 
\Bb ) \times \eb_z + \frac{B_0}{\tau_0}(q_\tn{h}\nhD\,\Ub - \jhD ) \times \eb_z + \tn{nonlinear\ terms\ for\ structure\ !}\,,  \label{lin1:U}
 \\[2mm]
 &\pder{p}{t} + \gamma\, p_0\, \nabla \cdot \Ub  = 0\,, 
 \\[2mm]
 & \pder{\Bb}{t} = B_0\,\nabla \times (\Ub \times \eb_z)\,.
 \\[3mm]
 & \pder{\fhD}{t} + \ub_\tn{g} \cdot \nabla \fhD + E_\tn{g}\,
\pder{\fhD}{v_\para} = 0\,,
\end{align}
$$

### Geometric structure with linear MHD part 

### Dispersion relation 

We linearize the MHD equations in the same way as before. Moreover, we take a DK equilibrium distribution of the form

$$
\begin{align}
 &f_{\tn{h}0}^\tn{D}=\left(\frac{m_\tn{h}}{2\pi}\right)^{3/2}\frac{n_{\tn{h}0}}{ \sqrt{T_{\tn{h}\parallel}}T_{\tn{h}\perp}}\exp\left(-\frac{m_\tn{h}q_\parallel^2}{2T_{\tn{h}\parallel}}\right)\exp\left(-\frac{m_\tn{h}\hat{\mu}B_0}{T_{\tn{h}\perp}}\right)\,,
\end{align}
$$

such that with $\int_{\hat{\mu}}\ldots\tn{d}q_\parallel=2\pi\int_0^\infty\int_{-\infty}^\infty\ldots\tn{d}q_\parallel\tn{d}\hat{\mu}$ we have

$$
\begin{align}
 &\int_{\hat{\mu}}\begin{pmatrix}1\\q_\parallel\\q_\parallel^2\\q_\parallel^3\end{pmatrix}f_{\tn{h}0}^\tn{D}B_0\,\tn{d}q_\parallel=\begin{pmatrix}n_{\tn{h}0}\\0\\n_{\tn{h}0}T_{\tn{h}\parallel}/m_\tn{h}\\0\end{pmatrix}\,,\qquad\int_{\hat{\mu}}\hat{\mu}f_{\tn{h}0}^\tn{D}B_0\,\tn{d}q_\parallel=\frac{n_{\tn{h}0}T_{\tn{h}\perp}}{m_\tn{h}B_0}\,.
\end{align}
$$

Let us start with the linearized DK equation:

$$
\begin{align}
 &\frac{\partial f_{\tn{h}1}^\tn{D}}{\partial t}+\ub_{\tn{g}0}\cdot\nabla f_{\tn{h}1}^\tn{D}=-E_{\tn{g}1}\frac{\partial f_{\tn{h}0}^\tn{D}}{\partial q_\parallel}\,.\\
 \Leftrightarrow\quad&\frac{\partial f_{\tn{h}1}^\tn{D}}{\partial t}+q_\parallel\eb_z\cdot\nabla f_{\tn{h}1}^\tn{D}=0\,,
\end{align}
$$

since $\ub_{\tn{g}0}=q_\parallel\eb_z$ and $E_{\tn{g}1}=0$. This equation simply describes an advection in $z-$ direction with the constant velocity $q_\parallel$. Thus, the solution is given by

$$
\begin{align}
 &f_{\tn{h}1}^\tn{D}(x,y,z,q_\parallel,\hat{\mu},t)=f_{\tn{h}1}^\tn{D}(x,y,z-q_\parallel t,q_\parallel,\hat{\mu},t=0):=f_{\tn{h}1}^\tn{D0}(x,y,z-q_\parallel t,q_\parallel,\hat{\mu})\,.
\end{align}
$$

For the computation of the perturbed current density $\jb_\tn{h1}^\tn{D}$, we need the linearized gyro-center velocity

$$
\begin{align}
 &\ub_\tn{g1}=\frac{q_\parallel}{B_0}\left(\Bb_1+\frac{q_\parallel}{a_\tn{h}B_0}\nabla\times\Bb_1\right)+\Ub_{1\perp}-\frac{q_\parallel}{a_\tn{h}B_0^2}\frac{\pa\Bb_1}{\pa t}\times\eb_z\,,\\
 \Rightarrow\quad&\int_\hat{\mu}\ub_\tn{g1}f_{\tn{h}0}^\tn{D}B_0\,\tn{d}q_\parallel=\frac{n_{\tn{h}0}T_{\tn{h}\parallel}}{a_\tn{h}B_0^2m_\tn{h}}\nabla\times\Bb_1+n_{\tn{h}0}\Ub_{1\perp}\,.
\end{align}
$$

The perturbed current density $\jb_\tn{h1}^\tn{D}$ is given by

$$
\begin{align}
 &\jb^\tn{D}_{\tn{h}1}=q_\tn{h}\int_{\hat{\mu}}(\ub_\tn{g0}f_{\tn{h}1}^\tn{D}+\ub_\tn{g1}f_{\tn{h}0}^\tn{D})B_0+\nabla\times\Mb_1=q_\tn{h}\eb_z\int_{\hat{\mu}}q_\parallel f_{\tn{h}1}^\tn{D0}B_0\,\tn{d}q_\parallel+\frac{q_\tn{h}n_{\tn{h}0}T_{\tn{h}\parallel}}{a_\tn{h}B_0^2m_\tn{h}}\nabla\times\Bb_1+q_\tn{h}n_{\tn{h}0}\Ub_{1\perp}+\nabla\times\Mb_1\,.
\end{align}
$$

Hence, we get for the linearized magnetization

$$
\begin{align}
 \Mb_1&=-m_\tn{h}\int_{\hat{\mu}}\hat{\mu}\Bb_1 f_{\tn{h}0}^\tn{D}\,\tn{d}q_\parallel-m_\tn{h}B_0\eb_z\int_{\hat{\mu}}\hat{\mu} f_{\tn{h}1}^\tn{D}\,\tn{d}q_\parallel+m_\tn{h}\int_{\hat{\mu}}\left(\frac{q_\parallel^3}{a_\tn{h}B_0^2}f_{\tn{h}0}^\tn{D}(\nabla\times\Bb_1)_\perp+q_\parallel f_{\tn{h}0}^\tn{D}\Ub_{1\perp}-\frac{q_\parallel^2}{a_\tn{h}B_0^2}f_{\tn{h}0}^\tn{D}\frac{\pa\Bb_1}{\pa t}\times\eb_z\right)\,\tn{d}q_\parallel\\
 &=-\frac{n_{\tn{h}0}T_{\tn{h}\perp}}{B_0^2}\Bb_1-m_\tn{h}B_0\eb_z\int_{\hat{\mu}}\hat{\mu} f_{\tn{h}1}^\tn{D0}\,\tn{d}q_\parallel-\frac{n_{\tn{h}0}T_{\tn{h}\parallel}}{a_\tn{h}B_0^3}\frac{\pa\Bb_1}{\pa t}\times\eb_z\,.
\end{align}
$$

This leads to the perturbed current density

$$
\begin{align}
 &\jb_{\tn{h}1}^\tn{D}=q_\tn{h}\eb_z\int_{\hat{\mu}}q_\parallel f_{\tn{h}1}^\tn{D0}B_0\,\tn{d}q_\parallel+\frac{q_\tn{h}n_{\tn{h}0}T_{\tn{h}\parallel}}{a_\tn{h}B_0^2m_\tn{h}}\nabla\times\Bb_1+q_\tn{h}n_{\tn{h}0}\Ub_{1\perp}-\frac{n_{\tn{h}0}T_{\tn{h}\perp}}{B_0^2}\nabla\times\Bb_1-\frac{n_{\tn{h}0}T_{\tn{h}\parallel}}{a_\tn{h}B_0^3}\nabla\times\left(\frac{\pa\Bb_1}{\pa t}\times\eb_z\right)-m_\tn{h}B_0\int_{\hat{\mu}}\hat{\mu}(\nabla\times\eb_z f_{\tn{h}1}^\tn{D0})\,\tn{d}q_\parallel\,.
\end{align}
$$

Let's assume a perturbed DK distribution function of the form $f_{\tn{h}1}^\tn{D}(z,q_\parallel,\hat{\mu},t)=\epsilon\cos(a(z-q_\parallel t))f_{\tn{h}0}^\tn{D}(q_\parallel,\hat{\mu})$ with $\epsilon\ll 1$.

$$
\begin{align}
 &\Rightarrow\quad q_\tn{h}\eb_z\int_{\hat{\mu}}q_\parallel f_{\tn{h}1}^\tn{D0}B_0\,\tn{d}q_\parallel=q_\tn{h}B_0\eb_z\epsilon\int_{\hat{\mu}}q_\parallel \cos(a(z-q_\parallel t))f_{\tn{h}0}^\tn{D}(q_\parallel,\hat{\mu})\,\tn{d}q_\parallel=\eb_z\frac{q_\tn{h}n_\tn{h}T_{\tn{h}\parallel}at}{m_\tn{h}}\exp\left(-\frac{T_{\tn{h}\parallel}a^2t^2}{2m_\tn{h}}\right)\sin(az)\,,\\
 &\Rightarrow\quad\jb_{\tn{h}1}^\tn{D}=\eb_z\frac{q_\tn{h}n_\tn{h}T_{\tn{h}\parallel}at}{m_\tn{h}}\exp\left(-\frac{T_{\tn{h}\parallel}a^2t^2}{2m_\tn{h}}\right)\sin(az)+\frac{q_\tn{h}n_{\tn{h}0}T_{\tn{h}\parallel}}{a_\tn{h}B_0^2m_\tn{h}}\nabla\times\Bb_1+q_\tn{h}n_{\tn{h}0}\Ub_{1\perp}-\frac{n_{\tn{h}0}T_{\tn{h}\perp}}{B_0^2}\nabla\times\Bb_1-\frac{n_{\tn{h}0}T_{\tn{h}\parallel}}{a_\tn{h}B_0^3}\nabla\times\left(\frac{\pa\Bb_1}{\pa t}\times\eb_z\right)
\end{align}
$$

In the momentum balance equation for the bulk velocity $\Ub$, we need the term

$$
\begin{align}
 &\frac{B_0}{\tau_0}\jb_\tn{h}^\tn{D}\times\eb_z=\frac{n_{\tn{h}0}}{\tau_0B_0}(T_{\tn{h}\parallel}-T_{\tn{h}\perp})(\nabla\times\Bb_1)\times\eb_z+\frac{B_0q_\tn{h}n_{\tn{h}0}}{\tau_0}\Ub_{1\perp}\times\eb_z-\frac{n_{\tn{h}0}T_{\tn{h}\parallel}}{\tau_0a_\tn{h}B_0^2}\left[\nabla\times\left(\frac{\pa\Bb_1}{\pa t}\times\eb_z\right)\right]\times\eb_z\,,
\end{align}
$$

By looking for plane wave solutions $\sim\exp[i(k_\parallel z-\omega t)]$, this results in 

$$
\begin{align}
 &\frac{B_0}{\tau_0}\hat{\jb}_\tn{h}^\tn{D}\times\eb_z=\frac{ik_\parallel n_{\tn{h}0}}{\tau_0B_0}(T_{\tn{h}\parallel}-T_{\tn{h}\perp})\begin{pmatrix}\hat{B}_x\\ \hat{B}_y\\0\end{pmatrix}+\frac{B_0q_\tn{h}n_{\tn{h}0}}{\tau_0}\begin{pmatrix}\hat{U}_y\\ -\hat{U}_x\\0\end{pmatrix}-\frac{k_\parallel\omega n_{\tn{h}0}T_{\tn{h}\parallel}}{\tau_0a_\tn{h}B_0^2}\begin{pmatrix}\hat{B}_y\\ -\hat{B}_x\\0\end{pmatrix}\,,
\end{align}
$$

which can be expressed only in terms of $\hat{\Ub}$ by using $\hat{\Bb}=-B_0k_\parallel\hat{\Ub}_\perp/\omega$ from Faraday's law:

$$
\begin{align}
 &\frac{B_0}{\tau_0}\hat{\jb}_\tn{h}^\tn{D}\times\eb_z=-\frac{ik_\parallel^2 n_{\tn{h}0}}{\tau_0}(T_{\tn{h}\parallel}-T_{\tn{h}\perp})\begin{pmatrix}\hat{U}_x\\ \hat{U}_y\\0\end{pmatrix}+\frac{B_0q_\tn{h}n_{\tn{h}0}}{\tau_0}\begin{pmatrix}\hat{U}_y\\ -\hat{U}_x\\0\end{pmatrix}+\frac{k_\parallel^2n_{\tn{h}0}T_{\tn{h}\parallel}}{\tau_0a_\tn{h}B_0}\begin{pmatrix}\hat{U}_y\\ -\hat{U}_x\\0\end{pmatrix}\,.
\end{align}
$$

This results in the the momentum balance equation

$$
\begin{align}
 &-i\omega\begin{pmatrix}\hat{U}_x\\\hat{U}_y\\\hat{U}_\parallel\end{pmatrix}+i\frac{v_\tn{s}^2k_\parallel^2}{\omega}\begin{pmatrix}0\\0\\\hat{U}_\parallel\end{pmatrix}+i\frac{v_\tn{A}^2k_\parallel^2}{\omega}\begin{pmatrix}\hat{U}_x\\\hat{U}_y\\0\end{pmatrix}-\frac{ik_\parallel^2 n_{\tn{h}0}}{\tau_0}(T_{\tn{h}\parallel}-T_{\tn{h}\perp})\begin{pmatrix}\hat{U}_x\\ \hat{U}_y\\0\end{pmatrix}+\frac{B_0q_\tn{h}n_{\tn{h}0}}{\tau_0}\begin{pmatrix}\hat{U}_y\\ -\hat{U}_x\\0\end{pmatrix}+\frac{k_\parallel^2n_{\tn{h}0}T_{\tn{h}\parallel}}{\tau_0a_\tn{h}B_0}\begin{pmatrix}\hat{U}_y\\ -\hat{U}_x\\0\end{pmatrix}-\frac{B_0q_\tn{h}n_{\tn{h}0}}{\tau_0}\begin{pmatrix}\hat{U}_y\\-\hat{U}_x\\0\end{pmatrix}=0\,.
\end{align}
$$

After multiplying by $i\omega$, replacing the temperatures by the thermal velocities $v_{\tn{th}\parallel}^2=2T_{\tn{h}\parallel}/m_\tn{h}$ respectively $v_{\tn{th}\perp}^2=2T_{\tn{h}\perp}/m_\tn{h}$ and using $m_\tn{i}=m_\tn{h}$, this is equivalent to

$$
\be
 \begin{pmatrix}
  \omega^2 - v_\tn{A}^2\,k_\para^2 + \nu_\tn{h}k_\para^2(v_{\tn{th}\parallel}^2-v_{\tn{th}\perp}^2)/2 & i\nu_\tn{h}\frac{\omega}{\Omega_{\tn{ci}}}k_\para^2v_{\tn{th}\parallel}^2/2 & 0 \\
  -i\nu_\tn{h}\frac{\omega}{\Omega_{\tn{ci}}}k_\para^2v_{\tn{th}\parallel}^2/2 & \omega^2 - v_\tn{A}^2\,k_\para^2 + \nu_\tn{h}k_\para^2(v_{\tn{th}\parallel}^2-v_{\tn{th}\perp}^2)/2 & 0 \\
  0 & 0 & \omega^2 - v_\tn{s}^2 \,k_\para^2
 \end{pmatrix}
 \begin{pmatrix}
  \hat{U}_x \\ \hat{U}_y \\ \hat{U}_\para
 \end{pmatrix}
 = 0\,,
\ee
$$

where $\nu_\tn{h}=n_{\tn{h}0}/n_0$ ($\tau_0=m_\tn{i}n_0$) is the ratio between the hot ion number density $n_{\tn{h}0}$ and the bulk density $n_0$ and $\Omega_\tn{ci}=q_\tn{i}B_0/m_\tn{i}$ is the ion cyclotron frequency. From this we end up with three kinds of eigenmodes. First, sound waves with $\omega=\pm v_\tn{s}k_\para$ which are unaffacted by the presence of energetic ions. Second, R-waves with 

$$
\begin{align}
 &D_R(k_\para,\omega)=\omega^2-v_\tn{A}^2k_\para^2+\nu_\tn{h}k_\para^2(v_{\tn{th}\parallel}^2-v_{\tn{th}\perp}^2)/2+\nu_\tn{h}\frac{\omega}{\Omega_{\tn{ci}}}k_\para^2v_{\tn{th}\parallel}^2/2=0\,,
\end{align}
$$

and third, L-waves with

$$
\begin{align}
 &D_L(k_\para,\omega)=\omega^2-v_\tn{A}^2k_\para^2+\nu_\tn{h}k_\para^2(v_{\tn{th}\parallel}^2-v_{\tn{th}\perp}^2)/2-\nu_\tn{h}\frac{\omega}{\Omega_{\tn{ci}}}k_\para^2v_{\tn{th}\parallel}^2/2=0\,.
\end{align}
$$

which finally gives

$$
\begin{align}
 &k_{\parallel,\tn{R/L}}=\pm\frac{\omega}{\sqrt{v_\tn{A}^2-\nu_\tn{h}(v_{\tn{th}\parallel}^2-v_{\tn{th}\perp}^2)/2-p_\tn{R/L}v_{\tn{th}\parallel}^2\omega/(2\Omega_\tn{ci})}}=\pm\frac{\omega}{v_\tn{A}\sqrt{1-\nu_\tn{h}\frac{v_{\tn{th}\parallel}^2}{2v_\tn{A}^2}(1-\frac{v_{\tn{th}\perp}^2}{v_{\tn{th}\parallel}^2}+p_\tn{R/L}\omega/\Omega_\tn{ci})}}\,,
\end{align}
$$

where $p_\tn{R/L}=\pm1$ to account for the polarization.

We can also solve for $\omega$ to arrive at

$$
\begin{align}
 &2\omega_{\tn{R/L},\pm}(k_\parallel)=-p_{\tn{R/L}}\nu_\tn{h}k_\parallel^2v_{\tn{th}\parallel}^2/(2\Omega_\tn{ci})\pm\sqrt{\nu_\tn{h}^2k_\parallel^4v_{\tn{th}\parallel}^4/(4\Omega_\tn{ci}^2)+4v_\tn{A}^2k_\parallel^2-2\nu_\tn{h}k_\parallel^2(v_{\tn{th}\parallel}^2-v_{\tn{th}\perp}^2)}\,.
\end{align}
$$

or in scaled variables $\omega^\prime=\omega/\Omega_\tn{ci}$, $v_{\tn{th}\parallel/\perp}^\prime=v_{\tn{th}\parallel/\perp}/v_\tn{A}$ and $k_\parallel^\prime=k_\parallel v_\tn{A}/\Omega_\tn{ci}$ 

$$
\begin{align}
 &2\omega_{\tn{R/L},\pm}^\prime(k_\parallel^\prime)=-p_{\tn{R/L}}\nu_\tn{h}k_\parallel^{\prime2}v_{\tn{th}\parallel}^{\prime2}/2\pm\sqrt{\nu_\tn{h}^2k_\parallel^{\prime4}v_{\tn{th}\parallel}^{\prime4}/4+4k_\parallel^{\prime2}-2\nu_\tn{h}k_\parallel^{\prime2}(v_{\tn{th}\parallel}^{\prime2}-v_{\tn{th}\perp}^{\prime2})}\,.
\end{align}
$$

In order to get wave growth or damping the square root must become negative. This means that

$$
\begin{align}
 &k_\parallel^{\prime 2}<\frac{4}{v_{\tn{th}\parallel}^{\prime2}\nu_\tn{h}^2}\left(\nu_\tn{h}(v_{\tn{th}\parallel}^{\prime2}-v_{\tn{th}\perp}^{\prime2})-2\right)\,.
\end{align}
$$


## Extended MHD with drift-kinetic equation

In extended MHD the electric field is given by

$$
\begin{align}
 & \Eb = - \Ub \times \Bb \left[ 1 - \frac{q_\tn{h} \nhD}{\tau}\,\left( a_\tn{i} 
- a_\tn{e} \right) \right]  + \frac{1}{\tau}\nabla \left( 
 a_\tn{e}\,p_\tn{i} - a_\tn{i}\,p_\tn{e}  \right) + \left( 
a_\tn{i} - a_\tn{e} \right) \frac{1}{\mu_0}\,\frac{(\nabla \times 
\Bb - \mu_0\,\jhD)}{\tau} \times\Bb\,,   \nonumber
\end{align}
$$

where $a_\tn{s} = q_\tn{s}/m_\tn{s}$ are constants and the last term is known as the _Hall term_. Extended MHD features two-fluid effects while having quasi-neutrality automatically built in. The equations read

\begin{align}
 &\pder{\tau}{t} + \nabla \cdot (\tau \Ub) = 0\,,
 \\[3mm]
 &\pder{\Ub}{t} + \Ub \cdot \nabla \Ub + \frac{\nabla (p_\tn{i} + 
 p_\tn{e})}{\tau}
  = - \frac{q_\tn{h}\nhD}{\tau}\,\Eb + \frac{1}{\mu_0}\,\frac{(\nabla \times 
\Bb - \mu_0\,\jhD)}{\tau} \times\Bb   \,,
 \\[3mm]
 &\pder{p_\tn{i}}{t} + \left[ \Ub \left( 1 + a_\tn{e}\,\frac{q_\tn{h} \nhD}{\tau} 
\right) + \frac{a_\tn{e}}{\mu_0}\frac{(\nabla \times \Bb - 
\mu_0\,\jhD) }{\tau} \right] 
\cdot \nabla 
p_\tn{i}
 \\[2mm]
 &\qquad + \gamma\, p_\tn{i} \nabla \cdot 
\left[ \Ub \left( 1 + a_\tn{e}\,\frac{q_\tn{h} \nhD}{\tau} \right) + 
\frac{a_\tn{e}}{\mu_0}\frac{(\nabla \times \Bb - \mu_0\,\jhD)} {\tau } \right]   = 0\,,  
\nonumber
 \\[4mm]
 &\pder{p_\tn{e}}{t} + \left[ \Ub \left( 1 - a_\tn{i}\,\frac{q_\tn{h} 
\nhD}{\tau} 
\right) - \frac{a_\tn{i}}{\mu_0}\frac{(\nabla \times \Bb - 
\mu_0\,\jhD)}{ \tau } 
\right] 
\cdot \nabla p_\tn{e}
 \\[2mm]
 &\qquad + \gamma\, p_\tn{e} \nabla \cdot \left[ \Ub \left( 1 - 
a_\tn{i}\,\frac{q_\tn{h} 
\nhD}{\tau} 
\right) - \frac{a_\tn{i}}{\mu_0}\frac{(\nabla \times \Bb - 
\mu_0\,\jhD)}{ \tau } 
\right] = 0 \,,   \nonumber
 \\[3mm]
 & \pder{\Bb}{t} = - \nabla \times \Eb\,,
 \\[3mm]
 & \pder{\fhD}{t} + \ub_\tn{g} \cdot \pder{\fhD}{\rb} + E_\tn{g}\,
\pder{\fhD}{q_\para} = 0\,.
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt


nuh = 5e-3
vth = 1.
v0  = 10.
Ah  = 1
Zh  = 1


def omega(k, pol, pm):
    return nuh*v0*Ah*k + pol*nuh*Ah**2*k**2/(4*Zh)*(vth**2 + 2*v0**2) + pm*k*np.sqrt(1 - v0**2*nuh*Ah*(1 - nuh*Ah) - pol*nuh*Ah**2*k/(2*Zh)*(3*v0*vth**2 + 2*v0**3 - nuh*v0*vth**2*Ah - 2*nuh*v0**3*Ah) + nuh**2*Ah**4*k**2/(16*Zh**2)*(vth**2 + 2*v0**2)**2)
```

```python
# full orbit dispersion relation
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt

def solveDispersionFullOrbit(k, pol, wch, vA, vth, v0, nuh, Ah, Zh, initial_guess, tol, max_it=100):
    
    
    def Z(xi):
        return np.sqrt(np.pi)*np.exp(-xi**2)*(1j - sp.erfi(xi))

    def Zprime(xi):
        return -2*(1 + xi*Z(xi))
    
    
    def D(k, w, pol):
        xi = (w - k*v0 + pol*wch)/(k*vth)
        
        return 1 - vA**2*k**2/w**2 + pol*nuh*wch/w + nuh*wch**2*Zh**2/(Ah*w**2)*(w - k*v0)/(k*vth)*Z(xi)
    
    def Dprime(k, w, pol):
        xi  = (w - k*v0 + pol*wch)/(k*vth)
        xip = 1/(k*vth)
        
        return 2*vA**2*k**2/w**3 - pol*nuh*wch/w**2 - 2*nuh*wch**2*Zh**2/(Ah*w**3)*(w - k*v0)/(k*vth)*Z(xi) + nuh*wch**2*Zh**2/(Ah*w**2)*1/(k*vth)*Z(xi) + nuh*wch**2*Zh**2/(Ah*w**2)*(w - k*v0)/(k*vth)*Zprime(xi)*xip
    
    w = initial_guess
    counter = 0
    
    while True:
        wnew = w - D(k, w, pol)/Dprime(k, w, pol)
        
        if np.abs(wnew - w) < tol or counter == max_it:
            w = wnew
            break

        w = wnew
        counter += 1

    return w, counter
```

```python
nuh = 5e-2
vth = 1.
v0  = 0.
Ah  = 1
Zh  = 1


k = np.linspace(0.5, 4., 100)


cases_v0 = 7
v0s = np.array([0., 0.5, 1.0, 1.5, 2., 2.5, 3.])

w1 = np.zeros((len(v0s), len(k)), dtype=complex)
w2 = np.zeros((len(v0s), len(k)), dtype=complex)

w1[:, 0] = 0.01 + 0j
w2[:, 0] = 0.01 + 0j

for j in range(len(v0s)):

    for i in range(len(k)):

        if i == 0:
            w1[j, i] = solveDispersionFullOrbit(k[i], +1, 1., 1., vth, v0s[j], nuh, Ah, Zh, w1[j, i], 1e-6, max_it=100)[0]
            w2[j, i] = solveDispersionFullOrbit(k[i], -1, 1., 1., vth, v0s[j], nuh, Ah, Zh, w2[j, i], 1e-6, max_it=100)[0]


        else:
            w1[j, i] = solveDispersionFullOrbit(k[i], +1, 1., 1., vth, v0s[j], nuh, Ah, Zh, w1[j, i - 1], 1e-6, max_it=100)[0]
            w2[j, i] = solveDispersionFullOrbit(k[i], -1, 1., 1., vth, v0s[j], nuh, Ah, Zh, w2[j, i - 1], 1e-6, max_it=100)[0]
```

```python
import psydac.core.interface as inter
import numpy as np
import sympy as sym
import scipy.sparse as sparse
import matplotlib.pyplot as plt
import derivatives as der
import projectors as proj
import evaluation as eva
import integrate as integr
import time
```

```python
# ... define boundary conditions (True: periodic, False: homogeneous Dirichlet)
bc = False

# ... length of computational domain in each direction
Lx = 1
Ly = 1
Lz = 1

# ... number of elements in each direction and corresponding element boundaries
Nel_x = 6
Nel_y = 6
Nel_z = 6

dx = Lx/Nel_x
dy = Ly/Nel_y
dz = Lz/Nel_z

el_b_x = np.linspace(0, Lx, Nel_x + 1)
el_b_y = np.linspace(0, Ly, Nel_y + 1)
el_b_z = np.linspace(0, Lz, Nel_z + 1)

# ... degree of spline basis (of the 0-forms) in each direction
p = [3, 3, 3]
px, py, pz = p

# ... number of basis functions and knot vectors
if bc == True:
    bcon = 1
    
    Nbase = [Nel_x + px, Nel_y + py, Nel_z + pz] 
    Nbase_x, Nbase_y, Nbase_z = Nbase
    
    Tx = inter.make_periodic_knots(px, Nbase_x)
    Ty = inter.make_periodic_knots(py, Nbase_y)
    Tz = inter.make_periodic_knots(pz, Nbase_z)
    T = [Tx, Ty, Tz]
    
    Nbase_x_0 = Nbase_x - px
    Nbase_y_0 = Nbase_y - py
    Nbase_z_0 = Nbase_z - pz

else:
    bcon = 0
    
    Nbase = [Nel_x + px, Nel_y + py, Nel_z + pz] 
    Nbase_x, Nbase_y, Nbase_z = Nbase
    
    Tx = inter.make_open_knots(px, Nbase_x)
    Ty = inter.make_open_knots(py, Nbase_y)
    Tz = inter.make_open_knots(pz, Nbase_z)
    T = [Tx, Ty, Tz]
    
    Nbase_x_0 = Nbase_x - 2
    Nbase_y_0 = Nbase_y - 2
    Nbase_z_0 = Nbase_z - 2



# ... discrete derivatives
GRAD = der.GRAD_3d(Nbase_x_0, Nbase_y_0, Nbase_z_0, bc)
CURL = der.CURL_3d(Nbase_x_0, Nbase_y_0, Nbase_z_0, bc)
DIV  = der.DIV_3d(Nbase_x_0, Nbase_y_0, Nbase_z_0, bc)

# ... mass matrices
M0 = integr.mass_matrix_V0(p, Nbase, T, bc)
M1 = integr.mass_matrix_V1(p, Nbase, T, bc)
M2 = integr.mass_matrix_V2(p, Nbase, T, bc)
M3 = integr.mass_matrix_V3(p, Nbase, T, bc)
```

```python
Psi_0 = lambda x, y, z : x*np.sin(2*np.pi*x)*y*(1 - np.cos(2*np.pi*y))*z*np.sin(2*np.pi*z)
rhs = integr.L2_prod_V0(Psi_0, p, Nbase, T)[1:-1, 1:-1, 1:-1]
```

```python
vec_0 = sparse.linalg.spsolve(M0, np.reshape(rhs, Nbase_x_0*Nbase_y_0*Nbase_z_0))
```
