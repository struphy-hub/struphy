# API: Dispersion relations

This API provides access to 1D dispersion relations available in Struphy.

```python
import numpy as np

from struphy.dispersion_relations import analytic

k = np.linspace(0, 0.3, 100)
```

## Maxwell

```python
disp_rel = analytic.Maxwell1D(c=2.0)
disp_rel.plot(k)
```

## MHD homogeneous slab

```python
disp_rel = analytic.MHDhomogenSlab()
disp_rel.plot(k)
```

## Hall MHD homogeneous slab

```python
disp_rel = analytic.ExtendedMHDhomogenSlab(eps=1.0)
disp_rel.plot(k)
```

## Fluid ITG in slab

```python
disp_rel = analytic.FluidSlabITG(vstar=0.1)
disp_rel.plot(k)
```
