# 1. Lie-Trotter

```python
import numpy as np
import matplotlib.pyplot as plt
import Utilitis_HybridCode as utils
```

```python
data1 = np.loadtxt('/home/florian/Schreibtisch/Masterarbeit/06_Github/learning-python/data_Nel=32_dt=0.0125_Np=1e5_T=200_amp=1e-4_p=1_FirstLie_NoCV')

en_B1  = data1[:, 0]
en_E1  = data1[:, 1]
en_C1  = data1[:, 2]
en_H1  = data1[:, 3]
```

```python
f1 = plt.figure()
f1.set_figheight(6)
f1.set_figwidth(6)

fontsize = 22
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)

en_tot1 = en_B1 + en_E1 + en_C1 + en_H1

Nt = len(en_B1) - 1
dt = 0.05/4
tn = np.linspace(0, Nt*dt, Nt + 1) 
tg = np.linspace(0, 120, 100)

colors = ['darkorange', 'purple', 'sienna', 'royalblue']

omega = utils.solveDispersionHybrid(2., +1, 1, -1, 2., 0.2, 0.53, 6e-2, 0.5 + 0j, 1e-8)[0]
omega_r = np.real(omega)
gamma = np.imag(omega)

plt.semilogy(tn, en_B1/en_tot1[0], linewidth = 2, label = r'$\mathcal{E}_{\tilde{B}}$', color = colors[0])
plt.semilogy(tn, en_E1/en_tot1[0], linewidth = 2, label = r'$\mathcal{E}_{\tilde{E}}$', color = colors[1])
plt.semilogy(tn, en_C1/en_tot1[0], '--', linewidth = 2, label = r'$\mathcal{E}_\mathrm{c}$', color = colors[2])
plt.semilogy(tn, en_H1/en_tot1[0], linewidth = 2, label = r'$\mathcal{E}_\mathrm{h}$', color = colors[3])
plt.semilogy(tg, 5e-6*np.exp(tg*gamma*2), 'k--', linewidth = 2)

plt.xlabel('$t|\Omega_\mathrm{ce}|$')
plt.ylabel(r'$\mathcal{E} / \mathcal{E}(0)$')
plt.title('Partition of energy (Lie-Trotter)', fontsize = fontsize)
plt.ylim((1e-8, 1e1))
plt.xlim((0, 200))
plt.legend(loc = 'lower right', ncol = 2)
plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Pr√§sentationen/05_Colloquium/01_Figures/GEMPIC_energies_LieTrotter.pdf', bbox_inches = 'tight')
```

```python
f2 = plt.figure()
f2.set_figheight(6)
f2.set_figwidth(6)

fontsize = 22
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)

en_tot1 = en_B1 + en_E1 + en_C1 + en_H1

Nt = len(en_B1) - 1
dt = 0.05/4
tn = np.linspace(0, Nt*dt, Nt + 1)

plt.semilogy(tn, np.abs(en_tot1 - en_tot1[0])/en_tot1[0], linewidth = 2, color = 'purple', label = 'FEEC (Lie-Trotter)')
plt.xlim((0, 200))
plt.ylim((1e-12, 1e0))
plt.xlabel('$t|\Omega_\mathrm{ce}|$')
plt.ylabel('$|\mathcal{E}(t) - \mathcal{E}(0)|/\mathcal{E}(0)$')
plt.title('Relative error in total energy', fontsize = fontsize)
plt.legend(fontsize = 18)
plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Pr√§sentationen/05_Colloquium/01_Figures/GEMPIC_error_LieTrotter.pdf', bbox_inches = 'tight')
```
