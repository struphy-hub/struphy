```python
import numpy as np
import matplotlib.pyplot as plt
import Utilitis_HybridCode as utils
from copy import deepcopy


fontsize = 22
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)

Lz = 1.
p = 1
Nel = 5

bsp, Nbase, quad_points, weights = utils.createBasis(Lz, Nel, p, bcs = 1)
el_b = np.linspace(0, Lz, Nel + 1) 

zplot = np.linspace(0, Lz, 200)

f1 = plt.figure()
f1.set_figheight(5)
f1.set_figwidth(6)

colors = []
color_counter = 0

for i in range(Nel):
    z = np.linspace(el_b[i], el_b[i + 1], 100)
    
    if i == 0:
        for jl in range(p + 1):
            j = i + jl
            pl = plt.plot(z, bsp(z, j), linewidth = 2)
            colors.append(pl[0].get_color())
        
    
    
    
    
    else:
        for jl in range(0, p + 1):
            j = i + jl
            
            if i <= Nel - p - 1:
            
                if j <  i + p:
                    plt.plot(z, bsp(z, j), linewidth = 2, color = colors[j])

                else:
                    pl = plt.plot(z, bsp(z, j), linewidth = 2)
                    colors.append(pl[0].get_color())
                    
            else:
            
                plt.plot(z, bsp(z, j), linewidth = 2, color = colors[j%Nel])

        
            
    
    if i <= Nel - 2 and i > 0:
        plt.plot(el_b[i]*np.ones(10), np.linspace(0, 0.75, 10), 'k--')



plt.plot(el_b[-2]*np.ones(10), np.linspace(0, 0.75, 10), 'k--', label = 'element boundaries')
#plt.plot(quad_points, np.zeros(len(quad_points)), 'ro', label = 'quadrature points')


plt.xlabel('$z$')
plt.ylim((0, 1.75))
plt.xlim((0, 1))
plt.legend()
plt.title('Periodic B-spline basis ($p=1$)', fontsize = fontsize)
plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Pr√§sentationen/05_Colloquium/01_Figures/Bsplines_p=1.pdf', bbox_inches = 'tight')
```
