# Dispersion relations for cold fluid/hybrid electrons

## 1. Cold fluid electrons

Only frequencies much larger than the ion plasma frequency $\Omega_\text{pi}$ are considered why ions are treated as a neutralizing stationary background. The remaining linear model for the perturbed electric field $\textbf{E}_1$, magnetic field $\textbf{B}_1$ and electron current density $\textbf{j}_1=qn_0\textbf{u}_1$ then reads (omit subscripts):

\begin{align}
\frac{\partial \textbf{j}}{\partial t} - \epsilon_0\Omega_\text{pe}^2\textbf{E} + \Omega_0\times\textbf{j} = 0, \\
\frac{1}{c^2}\frac{\partial \textbf{E}}{\partial t} - \nabla\times\textbf{B} + \mu_0\textbf{j} = 0, \\
\frac{\partial \textbf{B}}{\partial t} + \nabla\times\textbf{E} = 0,
\end{align}

with $\Omega_0=\frac{q\textbf{B}_0}{m}$ and $\Omega_\text{pe}^2=\frac{q^2 n_0}{\epsilon_0 m}$.

Assuming that all quantites ~$e^{i(kz-\omega t)}$ and $\textbf{B}_0=B_0\textbf{e}_z$ the following dispersion relations can be obtained for R- and L-waves, respectivley:

\begin{align}
D_\text{cold,R/L}(k,\omega)=1-\frac{c^2k^2}{\omega^2}-\frac{\Omega_\text{pe}^2}{\omega(\omega\pm\Omega_\text{ce})}=0 \Leftrightarrow k_\text{R/L} = \frac{\omega}{c}\sqrt{1-\frac{\Omega_\text{pe}^2}{\omega(\omega\pm\Omega_\text{ce})}}, 
\end{align}

where $\Omega_\text{ce}=\frac{qB_0}{m}=-\frac{eB_0}{m}<0$ for electrons.




## 2. Cold fluid electrons with artificial energetic particles

In order to get a damping/growth of the wave field, it is assumed that there is current density $\textbf{j}_\text{h}$ coming from energetic particles. For simplicity it is assumed that the there is linear response to the electric field of the from

\begin{align}
\textbf{j}_\text{h} = \sigma_\text{h}\textbf{E},
\end{align}

with some constant conductivity tensor

\begin{align}
\sigma_\text{h}=
\begin{pmatrix}
c_1 & c_2 &0 \\
-c_2 &c_1 &0 \\
0 &0 &c_3 \\
\end{pmatrix}.
\end{align}

The model then reads

\begin{align}
\frac{\partial \textbf{j}}{\partial t} - \epsilon_0\Omega_\text{pe}^2\textbf{E} + \Omega_0\times\textbf{j} = 0, \\
\frac{1}{c^2}\frac{\partial \textbf{E}}{\partial t} - \nabla\times\textbf{B} + \mu_0(\textbf{j}+\sigma_\text{h}\textbf{E}) = 0, \\
\frac{\partial \textbf{B}}{\partial t} + \nabla\times\textbf{E} = 0,
\end{align}

with the dispersion relation

\begin{align}
D_\text{art,R/L}(k,\omega)=D_\text{cold,R/L}(k,\omega)+\frac{c^2\mu_0}{\omega}(ic_1\mp c_2)
\end{align}

for R/L-waves ($c_3=0$).



## 3. Hybrid model
The self-consistend electron hybrid model with energetic electrons which are treated kinetically reads

\begin{align}
\frac{\partial \textbf{j}_\text{c}}{\partial t} = \epsilon_0\Omega_{\text{pe}}^2\textbf{E} + \Omega_\text{ce}\textbf{j}_\text{c}\times\textbf{e}_z,  \\
\frac{1}{c^2}\frac{\partial \textbf{E}}{\partial t} = \nabla\times\textbf{B} - \mu_0(\textbf{j}_\text{c} + \textbf{j}_\text{h}), \\
\frac{\partial \textbf{B}}{\partial t}=-\nabla\times\textbf{E}, \\
\frac{\partial f_\text{h}}{\partial t} + \textbf{v}\cdot\nabla f_\text{h} + \frac{q}{m}(\textbf{E}+\textbf{v}\times\textbf{B})\cdot\nabla_v f_\text{h}=0, \\
\textbf{j}_\text{h} = q\int\text{d}^3\textbf{v}\,\textbf{v}f_\text{h}.
\end{align}

Linear theory of the above model leads to the following general dispersion relation for an arbitrary equilibrium distribution function $f^0=f^0(v_\parallel,v_\bot)$:


\begin{align}
D_{\text{R/L}}(k,\omega)=1-\frac{c^2k^2}{\omega^2}-\frac{\Omega_{\text{pe}}^2}{\omega(\omega\pm\Omega_{\text{ce}})}+\nu_\text{h}\frac{\Omega_{\text{pe}}^2}{\omega}\int\text{d}^3\textbf{v}\frac{v_\bot}{2}\frac{\hat{G}f_\text{h}^0}{\omega\pm\Omega_{\text{ce}}-kv_\parallel}=0.
\end{align}

Here $\nu_\text{h}=n_\text{h}/n_\text{c}\ll1$ is the ratio between the hot and cold electron number densities, respectively, $\text{d}^3\textbf{v}=\text{d}v_\parallel\text{d}v_\bot v_\bot 2\pi$ and the differential operator

\begin{align}
\hat{G}=\frac{\partial}{\partial v_\bot}+\frac{k}{\omega}\left(v_\bot\frac{\partial}{\partial v_\parallel}-v_\parallel\frac{\partial}{\partial v_\bot}\right).
\end{align}

For an anisotropic Maxwellian 

\begin{align}
f^0(v_\parallel,v_\bot) = \frac{1}{(2\pi)^{3/2}w_\parallel w_\bot^2}\exp\left(-\frac{v_\parallel^2}{2w_\parallel^2}-\frac{v_\bot^2}{2w_\bot^2}\right)
\end{align}

the dispersion relation is given by

\begin{align}
D_{\text{R/L}}(k,\omega)=D_{\text{cold,R/L}}(k,\omega)+\nu_\text{h}\frac{\Omega_{\text{pe}}^2}{\omega^2}\left[\frac{\omega}{k\sqrt{2}w_\parallel}Z(\xi^{\pm})-\left(1-\frac{w_\bot^2}{w_\parallel^2}\right)(1+\xi^{\pm} Z(\xi^{\pm}))\right]=0, 
\end{align}

where $Z$ is the plasma dispersion function and 

\begin{align}
\xi^{\pm} = \frac{\omega\pm\Omega_\text{ce}}{k\sqrt{2}w_\parallel}.
\end{align}

# 1. Plasma dispersion function

```python
# ... parameters
wce = -1                           # ... electron cyclotron frequency
wpe = 2*np.abs(wce)                # ... cold plasma frequency (normalized to electron cyclotron frequency) 
wpar = 0.2                         # ... parallel thermal velocity of hot particles (normalized to speed of light)
wperp = 0.6                        # ... perpendiculer thermal velocity of hot particles (normalized to speed of light)
nuh = 5e-3                         # ... density ratio of cold/hot particles
mu0 = 1                            # ... vacuum permeability
c = 1                              # ... speed of light
# ...



# ... plasma dispersion function and its derivative
def Z(xi):
    return np.sqrt(np.pi)*np.exp(-xi**2)*(1j - sp.erfi(xi))

def Zprime(xi):
    return -2*(1 + xi*Z(xi))
# ...


fontsize = 22
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)


# ... create plots
f1 = plt.figure()
f1.set_figheight(4)
f1.set_figwidth(6)

xi = np.linspace(-4, 4, 100)
plt.plot(xi, np.real(Z(xi)), label = r'Re($Z(\xi)$)', linewidth = 2, color = 'darkorange')
plt.plot(xi, np.imag(Z(xi)), label = r'Im($Z(\xi)$)', linewidth = 2, color = 'purple')
plt.title(r'Plasma dispersion function $Z$', fontsize = fontsize)
plt.xlabel(r'$\xi$')
plt.ylim((-2, 2))
plt.xlim((-4, 4))
plt.legend()
#plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Prsentationen/05_Colloquium/01_Figures/Dispersion_Function.pdf', bbox_inches = 'tight')


f2 = plt.figure()
f2.set_figheight(4)
f2.set_figwidth(6)

plt.plot(xi, np.real(Zprime(xi)), label = r'Re($Z^\prime(\xi)$)', linewidth = 2, color = 'darkorange')
plt.plot(xi, np.imag(Zprime(xi)), label = r'Im($Z^\prime(\xi)$)', linewidth = 2, color = 'purple')
plt.title(r'Derivative of plasma dispersion function $\frac{\mathrm{d}Z}{\mathrm{d}\xi}$', fontsize = fontsize)
plt.xlabel(r'$\xi$')
plt.ylim((-2.5, 2.5))
plt.xlim((-4, 4))
plt.legend()
#plt.savefig('/home/florian/Schreibtisch/Masterarbeit/07_Ausarbeitung/master-thesis/01_Figures/03_HybridModel/Derivative_Dispersion_Function.pdf', bbox_inches = 'tight')
```

```python
# ... parameters
wce = -1                           # ... electron cyclotron frequency
wpe = 2*np.abs(wce)                # ... cold plasma frequency (normalized to electron cyclotron frequency) 
wpar = 0.2                         # ... parallel thermal velocity of hot particles (normalized to speed of light)
wperp = 0.6                        # ... perpendiculer thermal velocity of hot particles (normalized to speed of light)
nuh = 5e-3                         # ... density ratio of cold/hot particles
c = 1                              # ... speed of light
# ...




# ... Newton method for hybrid plasma dispersion relation for fixed k
k1 = np.linspace(0.4, 8.0, 100)

w1_1 = np.zeros(len(k1), dtype = complex)
w1_2 = np.zeros(len(k1), dtype = complex)
w1_3 = np.zeros(len(k1), dtype = complex)

wcut_R = 0.5*(np.sqrt(wce**2 + 4*wpe**2) - wce)
wcut_L = 0.5*(np.sqrt(wce**2 + 4*wpe**2) + wce)

w1_1[0] = 0.0001 + 0j
w1_2[0] = wcut_L + 0j
w1_3[0] = wcut_R + 0j

gamma_explicit = np.zeros(len(k1))


for i in range(0, len(k1)):
    
    if i == 0:
        w1_1[i] = utils.solveDispersionHybrid(k1[i], +1, c, wce, wpe, wpar, wperp, nuh, w1_1[i], 1e-6, 100)[0]
        w1_2[i] = utils.solveDispersionHybrid(k1[i], -1, c, wce, wpe, wpar, wperp, nuh, w1_2[i], 1e-6, 100)[0]
        w1_3[i] = utils.solveDispersionHybrid(k1[i], +1, c, wce, wpe, wpar, wperp, nuh, w1_3[i], 1e-6, 100)[0]
        
        gamma_explicit[i] = utils.solveDispersionHybridExplicit(k1[i], +1, c, wce, wpe, wpar, wperp, nuh, np.real(w1_1[i]), 1e-6, 1000)[1]
    
    else:
        w1_1[i] = utils.solveDispersionHybrid(k1[i], +1, c, wce, wpe, wpar, wperp, nuh, w1_1[i-1], 1e-6, 100)[0]
        w1_2[i] = utils.solveDispersionHybrid(k1[i], -1, c, wce, wpe, wpar, wperp, nuh, w1_2[i-1], 1e-6, 100)[0]
        w1_3[i] = utils.solveDispersionHybrid(k1[i], +1, c, wce, wpe, wpar, wperp, nuh, w1_3[i-1], 1e-6, 100)[0]
    
        gamma_explicit[i] = utils.solveDispersionHybridExplicit(k1[i], +1, c, wce, wpe, wpar, wperp, nuh, np.real(w1_1[i]), 1e-6, 1000)[1]
# ...


fontsize = 22
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)



f3 = plt.figure()
f3.set_figheight(4)
f3.set_figwidth(6)


plt.plot(k1, np.real(w1_3), 'darkorange', label = 'R', linewidth = 2)
plt.plot(k1, np.real(w1_2), 'purple', label = 'L', linewidth = 2)
plt.plot(k1, np.real(w1_1), 'sienna', label = 'R', linewidth = 2)

plt.ylim((0, 4))
plt.xlim((0, 8))
plt.plot(k1, np.ones(len(k1))*(-wce), 'k--', linewidth = 0.5, label = '$|\Omega_\mathrm{ce}|$')
plt.xlabel('$kc / |\Omega_\mathrm{ce}|$')
plt.ylabel('$\omega_\mathrm{r} / |\Omega_\mathrm{ce}|$')
plt.title('Real frequency', fontsize = fontsize)
plt.legend()
plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Prsentationen/05_Colloquium/01_Figures/Real_freq.eps', bbox_inches = 'tight')




f4 = plt.figure()
f4.set_figheight(4)
f4.set_figwidth(6)

plt.plot(k1, np.imag(w1_1), 'sienna', label = 'R', linewidth = 2)
plt.plot(np.linspace(-2, 10, len(k1)), np.zeros(len(k1)), 'k--', linewidth = 0.5)
plt.xlabel(r'$kc / |\Omega_\mathrm{ce}|$')
plt.ylabel(r'$\gamma / |\Omega_\mathrm{ce}|$')
plt.title('Growth rate', fontsize = fontsize)
plt.xlim((0, 8))
plt.legend()
plt.savefig('/home/florian/Schreibtisch/Masterarbeit/05_Prsentationen/05_Colloquium/01_Figures/Growth_rate.eps', bbox_inches = 'tight')
```

```python
# ... define figure layout
fontsize = 22
plt.rc('text', usetex = True)
plt.rc('font',family = 'serif')
plt.rcParams.update({'font.size': fontsize})
plt.rc('xtick', labelsize = fontsize)
plt.rc('ytick', labelsize = fontsize)
# ...

nuh6 = np.array([5e-3, 1e-2, 2e-2])
k6 = np.linspace(0.4, 8.0, 100)
wperp6 = 0.6

w6_1 = np.zeros((len(nuh6), len(k6)), dtype = complex)
w6_2 = np.zeros((len(nuh6), len(k6)), dtype = complex)
w6_3 = np.zeros((len(nuh6), len(k6)), dtype = complex)

w6_1[:, 0] = 0.0001 + 0j
w6_2[:, 0] = 2.0001 + 0j
w6_3[:, 0] = 2.5001 + 0j

gamma_explicit = np.zeros((len(nuh6), len(k6)))


for j in range(len(nuh6)):
    
    for i in range(0, len(k6)):
    
        if i == 0:
            w6_1[j, i] = utils.solveDispersionHybrid(k6[i], +1, c, wce, wpe, wpar, wperp6, nuh6[j], w6_1[j, i], 1e-6, 100)[0]
            w6_2[j, i] = utils.solveDispersionHybrid(k6[i], -1, c, wce, wpe, wpar, wperp6, nuh6[j], w6_2[j, i], 1e-6, 100)[0]
            w6_3[j, i] = utils.solveDispersionHybrid(k6[i], +1, c, wce, wpe, wpar, wperp6, nuh6[j], w6_3[j, i], 1e-6, 100)[0]
           
            gamma_explicit[j, i] = utils.solveDispersionHybridExplicit(k6[i], +1, c, wce, wpe, wpar, wperp6, nuh6[j], np.real(w6_1[j, i]), 1e-6, 1000)[1]

        else:
            w6_1[j, i] = utils.solveDispersionHybrid(k6[i], +1, c, wce, wpe, wpar, wperp6, nuh6[j], w6_1[j, i - 1], 1e-6, 100)[0]
            w6_2[j, i] = utils.solveDispersionHybrid(k6[i], -1, c, wce, wpe, wpar, wperp6, nuh6[j], w6_2[j, i - 1], 1e-6, 100)[0]
            w6_3[j, i] = utils.solveDispersionHybrid(k6[i], +1, c, wce, wpe, wpar, wperp6, nuh6[j], w6_3[j, i - 1], 1e-6, 100)[0]
                          
            gamma_explicit[j, i] = utils.solveDispersionHybridExplicit(k6[i], +1, c, wce, wpe, wpar, wperp6, nuh6[j], np.real(w6_1[j, i - 1]), 1e-6, 1000)[1]
            
f13 = plt.figure()
f13.set_figheight(4)
f13.set_figwidth(6)  

colors = ['darkorange', 'purple', 'sienna']
labels = [r'$\nu_\mathrm{h} = 2\,\%$', r'$\nu_\mathrm{h} = 1\,\%$', r'$\nu_\mathrm{h} = 0.5\,\%$', 'explicit']

for j in range(len(nuh6)):
    
    i = len(nuh6) - j - 1
    
    plt.plot(k6, np.imag(w6_1[i, :]), linewidth = 2, color = colors[j], label = labels[j])
    
    if j == len(nuh6) - 1:
        plt.plot(k6, gamma_explicit[i, :], 'k--', linewidth = 1, label = labels[-1])
    else:
        plt.plot(k6, gamma_explicit[i, :], 'k--', linewidth = 1)
        
    plt.plot(np.linspace(-2, 10, len(k6)), np.zeros(len(k6)), 'k--', linewidth = 0.5)
    plt.xlabel('$kc / |\Omega_\mathrm{ce}|$')
    plt.ylabel(r'$\gamma / |\Omega_\mathrm{ce}|$')
    plt.xlim((0, 8))
    plt.title('Comparison with explicit formula', fontsize = fontsize)
    plt.legend()

    

plt.savefig('/home/florian/Schreibtisch/Masterarbeit/07_Ausarbeitung/master-thesis/01_Figures/Comparison_explicit.pdf', bbox_inches = 'tight')
```
