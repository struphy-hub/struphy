# How to use .yml input files

```python
import yaml
```

```python
with open('test_input_dict.yml') as file:
    params = yaml.load(file)

    print(params)
```

```python
Nel          = params['Nel']
bc           = params['bc']
p            = params['p']
time_int     = params['time_int']
dt           = params['dt']
Tend         = params['Tend']
max_time     = params['max_time']
add_pressure = params['add_pressure']
kind_map     = params['kind_map']
params_map   = params['params_map']
gamma        = params['gamma']
add_PIC      = params['add_PIC']
Np           = params['Np']
control      = params['control']
v0           = params['v0']                    
vth          = params['vth']

v0x          = v0[0]
v0y          = v0[1]                       
v0z          = v0[2]

print('Nel:'.rjust(13), Nel)
print('bc:'.rjust(13), bc)
print('p:'.rjust(13), p)
print('time_int:'.rjust(13), time_int)
print('dt:'.rjust(13), dt)
print('Tend:'.rjust(13), Tend)
print('max_time:'.rjust(13), max_time)
print('add_pressure:'.rjust(13), add_pressure)
print('kind_map:'.rjust(13), kind_map)
print('params_map:'.rjust(13), params_map)
print('gamma:'.rjust(13), gamma)
print('add_PIC:'.rjust(13), add_PIC)
print('Np:'.rjust(13), Np)
print('control:'.rjust(13), control)
print('v0x:'.rjust(13), v0x)
print('v0y:'.rjust(13), v0y)
print('v0z:'.rjust(13), v0z)
print('vth:'.rjust(13), vth)
```
