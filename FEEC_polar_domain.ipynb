# FEEC on domains with a polar singularity

## Preliminaries

The FEEC formalism runs into severe troubles when the mapping $F:U\to\Omega$ from the logical domain to the physical domain has a polar singularity. Let us consider the generic example of polar coordinates in 2D and let $(r,\theta) \in U$ denote the logical (polar) coordinates on the rectangle $U=[0,1]\times [0,2\pi)$, whereas $(x,y) \in \Omega$ stand for the physical coordinates on the unit disc $\Omega = S^1$:

$$
\be
 F: U\to\Omega\,,\qquad  (r,\theta) \mapsto (x,y)\,, \qquad \begin{pmatrix} x \\ y \end{pmatrix} = 
 \begin{pmatrix} r \cos\theta \\ r\sin\theta \end{pmatrix}\,.
\ee
$$

The Jacobian and its inverse are given by

$$
\be
 DF = \begin{pmatrix}
  \pder{x}{r} & \pder{x}{\theta}
  \\
  \pder{y}{r} & \pder{y}{\theta}
 \end{pmatrix} =
 \begin{pmatrix}
  \cos\theta & -r\sin\theta
  \\
  \sin\theta & r \cos\theta
 \end{pmatrix}\,, \qquad
 DF^{-1} = \frac{1}{r}\begin{pmatrix}
  r\cos\theta & r\sin\theta
  \\
  - \sin\theta & \cos\theta
 \end{pmatrix} \,,
\ee
$$

from which follow the metric tensor and related quantities

$$
\be
 G = DF^\top DF = \begin{pmatrix}
 1 & 0 \\ 0 & r^2
 \end{pmatrix}\,,\qquad g = \det G = r^2\,,\qquad G^{-1} = \begin{pmatrix}
 1 & 0 \\ 0 & \frac{1}{r^2}
 \end{pmatrix}\,.
\ee
$$

Moreover, the basic 1-forms $dr$ and $d\theta$, which are the lines of $(DF)^{-1}$, read

$$
\be
 dr = \begin{pmatrix} \cos\theta \\ \sin\theta  \end{pmatrix}\,,\qquad d\theta = \frac{1}{r} \begin{pmatrix}   -\sin\theta \\ \cos\theta  \end{pmatrix}\,.
\ee
$$

The whole FEEC formalism breaks down at $r = 0$ because $d\theta$ is not defined at that point. Ignoring this fact, we still can construct the following 2D deRham complexes on the logical domain $U$:

where

$$
\ba
 H(\curl) &:= \{ \Eb \in (L^2(U))^2: \ \curl\, \Eb \in L^2(U) \}\,,
 \\[2mm]
 H(\div) &:= \{ \Eb \in (L^2(U))^2: \ \div\, \Eb \in L^2(U) \}\,,
\ea
$$

and the involved operators are defined, for $\phi:U\to\RR$ and $\Eb = (E_r,E_\theta): U\to\RR^2$, as

$$
\ba
 & \grad\, \phi := \begin{pmatrix} \pa_r \phi \\ \pa_\theta \phi \end{pmatrix}\,,\qquad \curl\, \Eb := \pa_r E_\theta - \pa_\theta E_r\,,  \label{def:grad}
 \\[2mm]
 &\grad^\perp\, \phi := \begin{pmatrix} \pa_\theta \phi \\ -\pa_r \phi \end{pmatrix}\,,\qquad \div\, \Eb := \pa_r E_r + \pa_\theta E_\theta  \,.
\ea
$$

Clearly $\curl\, \grad = 0$ and $\div\,\grad^\perp = 0$ as necessary for a complex. Moreover, it is not hard to see that the adjoint of $\grad$ is $-\div$ and the adjoint of $\grad^\perp$ is $\curl$ (assuming the boundary terms vanish). We can write down the weak form of the Poisson equation in terms of differential forms using the left complex from above: find $(E^1,\phi^0) \in H\Lambda^1 \times H\Lambda^0$ such that

$$
\begin{alignat}{3}
&(E^1,F^1) + (\dd \phi^0,F^1) &&= 0\qquad &&\forall\ F^1 \in H\Lambda^1
 \\[1mm]
 &-(E^1, \dd \psi^0) &&= (*n^2, \psi^0) \qquad &&\forall\ \psi^0 \in H\Lambda^0\,.
\end{alignat}
$$

This is coordinate-independent. From the definition of the spaces of $p$-forms,

$$
\begin{align}
 L^2 \Lambda^p(\Omega) &:= \left\{ \alpha^p \in \Lambda^p(\Omega)\ : \quad (\alpha^p,\,\alpha^p) < \infty \right\}\,,
 \\[2mm]
 H \Lambda^p(\Omega) &:= \left\{ \alpha^p \in L^2\Lambda^p(\Omega)\,,\ \tn d \alpha^p \in L^2 \Lambda^{p+1}(\Omega) \right\}\,,
\end{align}
$$

we see that all the above integrals are well-defined. Thus, for any 1-form $E^1 \in H\Lambda^1$ with components $E^1 = E_r\, dr + E_\theta\, d\theta$, $\Eb = (E_r,E_\theta)$, on the logical domain we have

$$
\be \label{singint}
 (E^1,E^1) = \int_U \Eb^\top G^{-1} \Eb \sqrt g \,dr d\theta
 =
 \int_U \Eb^\top \begin{pmatrix}
 r & 0 \\ 0 & \frac{1}{r}
 \end{pmatrix} \Eb \,dr d\theta = \int_U \left(r\, E_r^2 + \frac{E_\theta^2}{r} \right) \,dr d\theta < \infty\,.
\ee
$$

This implies that

$$
\be \label{limgrad}
 E^1 \in H\Lambda^1 \qquad \Rightarrow \qquad \lim_{r\to 0} \int_0^{2\pi} E_\theta^2(r,\theta)\, d\theta = \int_0^{2\pi} \lim_{r\to 0} E_\theta^2(r,\theta) \, d\theta = 0 \qquad \Rightarrow \qquad \lim_{r\to 0} E_\theta(r,\theta) = 0\,.
\ee
$$

On the logical domain the above Poisson system reads

$$
\be \label{Poisson:df:weak}
\left\{
\begin{aligned}
& \Eb + \nabla \phi = 0\qquad \tn{in}\ H(\curl,U)\,,
 \\[1mm]
 -\int_U &\Eb^\top G^{-1} \nabla \psi \sqrt{g} \,dr d\theta  = o\int_U n\, \psi\,dr d\theta \qquad \forall\ \psi \in H^1(U)\,,
\end{aligned}
\right.
\ee
$$

where $\nabla = \grad$ as defined in \eqref{def:grad}.

## $C^0$-continuous deRham sequence for polar domains

The recipe for constructing $C^m$-continuous basis functions at the pole can be found in 

* D. Toshniwal et al., _Multi-degree smooth polar splines: A framework for geometric modeling and isogeometric analysis_, Comput. Methods Appl. Mech. Engrg. 316 (2017) 1005â€“1061. 

We follow their ideas and aim to construct $C^0$-continuity at the pole, in all spaces $V_0$, $V_1$ and $V_2$. In fact, we will construct $C^1$-continuity in $V_0$ and $C^0$-continuity in the other two spaces. Suppose we have a B-spline basis of degree $k$ on the $r$-domain $[0, r_\tn{max}]$, denoted $(N_{i_1}^k(r))_{i_1=0}^{n_1-1}$, and a B-spline basis of degree $k$ on the $\theta$-domain $[0,2\pi)$, denoted $(N_{i_2}^k(\theta))_{i_2=0}^{n_2-1}$. We could also take different degrees $k_1$ and $k_2$ for $r$ and $\theta$ but choose the same degree in both directions for convenience. We demand the B-splines to be clamped at $r=0$, i.e.

$$
\be
 N_{i_1}^k(0) = 
 \begin{cases}
  1\,, & \ \tn{if} \ i_1 = 0\,,
  \\[1mm]
  0\,, & \ \tn{if} \ i_1 >0\,.
 \end{cases}
\ee
$$

The standard tensor product basis of $V_0 \subset H^1(U)$ with multi-index $\ib = (i_1,i_2)$ would read $\Lambda_\ib^0 = N_{i_1}^k(r) N_{i_2}^k(\theta)$. Note that the degrees of freedom (DOFs) with index $i_1 = 0$ are mapped to the pole in the physical domain. Following Toshniwal et al., a basis that is $C^m$-continuous at the pole $r=0$ can be constructed from linear combinations of the basis functions that correspond to DOFs $i_1 \leq m$. In case of $m=1$ there will be three new basis functions instead of the $2 n_2$ tensor product basis functions used in the linear combinations. Therefore, the dimension of the space $V_0$ is $\tn{dim} V_0 = n_1n_2 - 2 n_2 + 3$. The three new basis functions will be denoted $\chi^0_l$, $l=1,2,3$, and written as

$$
\be
 \chi^0_l(r,\theta) = \sum_{i_1 = 0}^1 \sum_{i_2 = 0}^{n_2-1} c^l_{i_1,i_2}  N_{i_1}^k(r) N_{i_2}^k(\theta)\,,\qquad l=1,2,3\,.
\ee
$$

Here, according to Toshniwal et al., the coefficients must be chosen as

$$
\be
 c^l_{0,i_2} = \frac{1}{3} \quad \forall \ i_2 \qquad \tn{and} \qquad 
 \begin{pmatrix}
 c^1_{1,i_2}
 \\
 c^2_{1,i_2}
 \\
 c^3_{1,i_2}
 \end{pmatrix} =
 \begin{pmatrix}
 \frac{1}{3} & 0 & \frac{1}{3}
 \\
 -\frac{1}{6} & \frac{1}{2\sqrt 3} & \frac{1}{3}
 \\
 -\frac{1}{6} & -\frac{1}{2\sqrt 3} & \frac{1}{3}
 \end{pmatrix}
 \begin{pmatrix}
 \cos \theta_{i_2}
 \\
 \sin \theta_{i_2}
 \\
 1
 \end{pmatrix}\,.
\ee
$$


Due to the partition-of-unity property of the B-splines we have

$$
\be
 \sum_{i_2=0}^{n_2-1} N_0^k(r) N_{i_2}^k(\theta) = N_0^k(r)\,,
\ee
$$

such that the $C^1$-continuous basis for $V_0 \subset H^1(U)$ reads

$$
\be \label{basis:V0}
 \Lambda^0_\ib(r,\theta) := \begin{cases}
 \chi^0_l(r,\theta) = \frac{1}{3} N_0^k(r) + \sum_{i_2 = 0}^{n_2-1} c^l_{1,i_2}  N_{1}^k(r) N_{i_2}^k(\theta)\,, & \ \tn{for} \ l = 1,2,3
 \\[2mm]
 N_{i_1}^k(r) N_{i_2}^k(\theta)\,, & \ \tn{for} \ i_1 \geq 2\,,
 \end{cases}
\ee
$$

Taking the exterior derivative of these Basis functions leads to the following basis for $V_1 \subset H(\curl,U)$,

$$
\ba
 \tn dr:\qquad \Lambda^1_{\ib,1}(r,\theta) &:= \begin{cases}
  -\frac{1}{3} D_1^k(r) + \sum_{i_2 = 0}^{n_2-1} c^l_{1,i_2}  D_1^k(r)  N_{i_2}^k(\theta)\,, & \ \tn{for} \ l = 1,2,3
 \\[1mm]
 D_{i_1}^k(r) N_{i_2}^k(\theta)\,, & \ \tn{for} \ i_1 \geq 2\,,
 \end{cases}  \label{basis:V1:r}
 \\[3mm]
 \tn d \theta:\qquad \Lambda^1_{\ib,2}(r,\theta) &:= \begin{cases}
 \sum_{i_2 = 0}^{n_2-1} \left( c^l_{1,i_2} - c^l_{1,i_2-1} \right)  N_{1}^k(r) D_{i_2}^k(\theta) \,, & \ \tn{for} \ l = 1,2,3
 \\[1mm]
 N_{i_1}^k(r) D_{i_2}^k(\theta)\,, & \ \tn{for} \ i_1 \geq 2\,.
 \end{cases}  \label{basis:V1:theta}
\ea
$$

In particular, we remark that 

$$
\be
 \lim_{r\to 0} \Lambda^1_{\ib,2}(r,\theta) = 0 \qquad \forall \ \ib\,,
\ee
$$

such that the important property \eqref{limgrad} is satisfied by all basis functions of $V_1$. Finally, by taking the exterior derivative of the basis functions \eqref{basis:V1:r} and \eqref{basis:V1:theta}, which corresponds to taking the $\curl$, we obtain the basis for $V_2 \subset L^2(U)$:

$$
\be \label{basis:V2}
 \Lambda^2_\ib(r,\theta) := \begin{cases}
 \sum_{i_2 = 0}^{n_2-1} \left( c^l_{1,i_2} - c^l_{1,i_2-1} \right)  D_{1}^k(r) D_{i_2}^k(\theta) \,, & \ \tn{for} \ l = 1,2,3
 \\[1mm]
 D_{i_1}^k(r) D_{i_2}^k(\theta)\,, & \ \tn{for} \ i_1 \geq 2\,.
 \end{cases}
\ee
$$

__Proposition 1. (Exact sequence property.)__ The 2D-$H(\curl)$-diagram satisfies the exact sequence property

$$
\be
 \Im\, \grad V_0 = \tn{Ker}\, \curl V_1
\ee
$$

if we choose the above bases for $V_0$, $V_1$ and $V_2$.

_Proof:_ This follows from the fact that $\grad\, \Lambda_\ib^0 \in V_1$ and $\curl\, \Lambda_\ib^1 \in V_2$ by construction, and from $\curl\,\grad = 0$.



## Projectors

We now want to ensure the commuting property of the diagram with the new basis, which is not a tensor product basis anymore. For this let us introduce the following points in the logical domain $U$:

$$
\ba
 3\ \tn{radii:}\qquad &r^*_1 \neq r^*_2 \neq r^*_3 \,\quad\in (0,r_2)\,,
 \\[2mm]
 3\ \tn{angles:}\qquad &\theta^*_1 \neq \theta^*_2 \quad \in [0,2\pi) \quad \tn{and} \quad \theta^*_3  \in \{\theta_0,\ldots,\theta_{n_2-1} \}\,.
\ea
$$

For $a \in H^1$ we have $(\Pi_0 a) \in V_0$, expressed in the basis \eqref{basis:V0} as

$$
\be
 (\Pi_0 a)(r,\theta) = \sum_{l=1}^3 a_l\, \chi^0_l(r,\theta) + \sum_{i_1 =2}^{n_1-1}\sum_{i_2 = 0}^{n_2-1} a_{i_1,i_2}\, N_{i_1}^k(r) N_{i_2}^k(\theta)\,.
\ee
$$

We define the coefficients $(a_1,a_2,a_3,a_{i_1,i_2}) \in \RR^{N}$, with $N=n_1n_2-2n_2+3$ as follows: recalling that $r_0 = 0$, we demand

$$
\be \label{def:Pi0}
 \Pi_0:\ \begin{cases}
 (\Pi_0 a)(0) = \frac{1}{2}(a_1 + a_2 + a_3) = \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} a(0,\theta_{j_2})\,, 
 \\[2.5mm]
 (\Pi_0 a)(r_1,\theta_n^*) = a(r_1,\theta_n^*)\,, & \ \tn{for} \ n = 1,2\,, 
 \\[3mm]
 (\Pi_0 a)(r_{j_1}, \theta_{j_2}) = a(r_{j_1}, \theta_{j_2})\,, & \ \tn{for} \ j_1 \geq 2\,.
 \end{cases}
\ee
$$

For $\Eb=(E_r,E_\theta) \in H(\curl,U)$ we define the $2N$ coefficients of $(\Pi_1 \Eb) \in V_1$ as follows: for the first component $(\Pi_1 \Eb)_r$, we demand

$$
\be \label{def:Pi1:1}
 \Pi_{1,r}:\ \begin{cases}
 \int_0^{r_1}(\Pi_1 \Eb)_r(r,\theta_n^*)\,\tn d r = \mathring{E_r}(r_1, \theta_n^*) - \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} \mathring{E_r}(0, \theta_{j_2}) \,, & \ \tn{for} \ n = 1,2\,, 
 \\[3mm]
 \int_0^{r_2}(\Pi_1 \Eb)_r(r,\theta_3^*)\,\tn d r = \mathring{E_r}(r_2, \theta_3^*) - \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} \mathring{E_r}(0, \theta_{j_2}) \,,
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}}(\Pi_1 \Eb)_r(r,\theta_{j_2})\,\tn d r = \int_{r_{j_1}}^{r_{j_1+1}}E_r(r,\theta_{j_2})\,\tn d r \,, & \ \tn{for} \ j_1 \geq 2\,,
 \end{cases}
\ee
$$

where $\mathring{E_r}$ denotes the $r$-primitive of the first component $E_r$ of $\Eb$, 

$$
\be \label{def:prim}
 \mathring{E_r}(r,\theta) := \int_0^r E_r(s,\theta)\,\tn d s - \int_0^{r_\tn{max}} E_r(s,\theta)\,\tn d s\,.
\ee
$$

Remark that the constant of integration in this primitive integral has been set to the $r$-average over the domain $[0,r_\tn{max}]$. This will be sufficient to achieve the commuting property. For the second component $(\Pi_1 \Eb)_\theta$ of $\Pi_1 \Eb$ demand

$$
\be \label{def:Pi1:2}
 \Pi_{1,\theta}:\ \begin{cases}
 \int_{0}^{2\pi} (\Pi_1 \Eb)_\theta(r^*_n,\theta)\,\tn d \theta = \int_{0}^{2\pi} E_\theta(r^*_n,\theta)\,\tn d \theta\,, & \ \tn{for} \ n = 1,2,3\,, 
 \\[3mm]
 \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_1 \Eb)_\theta(r_{j_1},\theta)\,\tn d \theta = \int_{\theta_{j_2}}^{\theta_{j_2+1}}E_\theta(r_{j_1},\theta)\,\tn d \theta \,, & \ \tn{for}\ \ j_1 \geq 2 \,.
 \end{cases}
\ee
$$

Finally, for $a \in L^2(U)$ we define the $N$ coefficients of $\Pi_2 a \in V_2$ via

$$
\be \label{def:Pi2}
 \Pi_2: \ \begin{cases}
 \int_0^{r^*_n} \int_0^{2\pi} (\Pi_2 a)(r,\theta)\,\tn dr \tn d\theta = \int_0^{r^*_n} \int_0^{2\pi} a(r,\theta)\,\tn dr \tn d\theta\,, \ & \tn{for} \ n=1,2,3\,,
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_2 a)(r,\theta)\,\tn d r\tn d \theta =    \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}} a(r,\theta)\,\tn d r\tn d \theta\,,\ & \tn{for} \ j_1 \geq 2\,. 
 \end{cases}
\ee
$$


__Proposition 2. (Commuting diagram for polar domain.)__ The 2D-$H(\curl)$-diagram commutes if

* we choose as basis functions for $V_0$, $V_1$ and $V_2$ the functions defined in \eqref{basis:V0}, \eqref{basis:V1:r}-\eqref{basis:V1:theta} and \eqref{basis:V2}, respectively, 
* we define the projector $\Pi_0$ as in \eqref{def:Pi0}, the projector $\Pi_1$ as in \eqref{def:Pi1:1}-\eqref{def:Pi1:2} and the projector $\Pi_2$ as in \eqref{def:Pi2}, 
* we impose Dirichlet boundary conditions independent of $\theta$ at the outer boundary $r=r_\tn{max}$, i.e. for $\psi \in H^1(U)$ we demand

$$
\be \label{bc:Dirichlet}
 \psi(r_\tn{max},\theta) = c \in \RR\,.
\ee
$$

_Proof:_ For $\psi \in H^1$ we have $(\Pi_0\psi) \in V_0$ and $\grad (\Pi_0\psi) = (\pa_r (\Pi_0\psi),\, \pa_\theta (\Pi_0\psi))^\top \in V_1$ by construction (see Proposition 1). Then, 

$$
\be \label{diag:1}
 \begin{cases}
 \int_0^{r_1}\pa_r (\Pi_0\psi)(r,\theta_n^*)\,\tn d r = \Pi_0\psi(r_1, \theta_n^*) -  \Pi_0\psi(0, \theta_n^*) \stackrel{\eqref{def:Pi0}}{=} \psi(r_1, \theta_n^*) - \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} \psi(0, \theta_{j_2}) \,, & \ \tn{for} \ n = 1,2\,, 
 \\[3mm]
 \int_0^{r_2}\pa_r (\Pi_0\psi)(r,\theta_3^*)\,\tn d r = \Pi_0\psi(r_2, \theta_3^*) -  \Pi_0\psi(0, \theta_3^*) \stackrel{\eqref{def:Pi0}}{=} \psi(r_2, \theta_3^*) - \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} \psi(0, \theta_{j_2}) \,,
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}}\pa_r (\Pi_0\psi)(r,\theta_{j_2})\,\tn d r = \Pi_0\psi(r_{j_1+1}, \theta_{j_2}) -   \Pi_0\psi(r_{j_1}, \theta_{j_2}) \stackrel{\eqref{def:Pi0}}{=} \psi(r_{j_1+1}, \theta_{j_2}) - \psi(r_{j_1}, \theta_{j_2})  \,, & \ \tn{for} \ j_1 \geq 2\,.
 \end{cases}
\ee
$$

On the other hand, using the Dirichlet boundary condition \eqref{bc:Dirichlet} in the primitive

$$
\be
\begin{aligned}
 \mathring{(\pa_r\psi)}(r,\theta) &= \int_0^r \pa_r\psi(s,\theta)\,\tn d s - \int_0^1 \pa_r\psi(s,\theta)\,\tn d s
 \\[1mm]
 &= \psi(r,\theta) - \psi(0,\theta) - \psi(1,\theta) + \psi(0,\theta) 
 \\[3mm]
 &= \psi(r,\theta) - c\,,
\end{aligned}
\ee
$$

for the $r$-component of $\Pi_1 \grad\, \psi$ we obtain

$$
\be \label{diag:2}
 \begin{cases}
 \int_0^{r_1}(\Pi_1 \grad\, \psi)_r(r,\theta_n^*)\,\tn d r \stackrel{\eqref{def:Pi1:1}}{=} \psi(r_1, \theta_n^*) - \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} \psi(0, \theta_{j_2}) \,, & \ \tn{for} \ n = 1,2\,, 
 \\[3mm]
 \int_0^{r_2}(\Pi_1 \grad\, \psi)_r(r,\theta_3^*)\,\tn d r \stackrel{\eqref{def:Pi1:1}}{=} \psi(r_2, \theta_3^*) - \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} \psi(0, \theta_{j_2}) \,,
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}}(\Pi_1 \grad\, \psi)_r(r,\theta_{j_2})\,\tn d r \stackrel{\eqref{def:Pi1:1}}{=} \psi(r_{j_1+1},\theta_{j_2}) - \psi(r_{j_1},\theta_{j_2}) \,, & \ \tn{for} \ j_1 \geq 2\,.
 \end{cases}
\ee
$$

For the second component we have

$$
\be \label{diag:3}
 \begin{cases}
 \int_{0}^{2\pi} \pa_\theta (\Pi_0\psi)(r^*_n,\theta)\,\tn d \theta = 0\,, & \ \tn{for} \ n = 1,2,3\,, 
 \\[3mm]
 \int_{\theta_{j_2}}^{\theta_{j_2+1}}\pa_\theta (\Pi_0\psi)(r_{j_1},\theta)\,\tn d \theta = (\Pi_0\psi)(r_{j_1},\theta_{j_2+1}) - (\Pi_0\psi)(r_{j_1},\theta_{j_2}) \stackrel{\eqref{def:Pi0}}{=} \psi(r_{j_1},\theta_{j_2+1}) - \psi(r_{j_1},\theta_{j_2}) \,, & \ \tn{for}\ \ j_1 \geq 2 \,.
 \end{cases}
\ee
$$

Moreover, for the $\theta$-component of $\Pi_1 \grad\, \psi$ we obtain

$$
\be \label{diag:4}
 \begin{cases}
 \int_{0}^{2\pi} (\Pi_1 \grad\, \psi)_\theta(r^*_n,\theta)\,\tn d \theta \stackrel{\eqref{def:Pi1:2}}{=} 0 \,, & \ \tn{for} \ n = 1,2,3\,, 
 \\[3mm]
 \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_1 \grad\, \psi)_\theta(r_{j_1},\theta)\,\tn d \theta \stackrel{\eqref{def:Pi1:2}}{=} \psi(r_{j_1},\theta_{j_2+1}) - \psi(r_{j_1},\theta_{j_2}) \,, \ & \tn{for}\ \ j_1 \geq 2 \,.
 \end{cases}
\ee
$$

As \eqref{diag:1} is equal to \eqref{diag:2} and also \eqref{diag:3} is equal to \eqref{diag:4}, we have proved that the left part of the diagram commutes:

$$
\be
 \Pi_1 \grad = \grad\,\Pi_0\,.
\ee
$$

Let us continue with the right part of the diagram. For $\Eb \in H(\curl)$ we have $(\Pi_1\Eb) \in V_1$ and $\curl (\Pi_1\Eb) = (\pa_r (\Pi_1\Eb)_\theta - \pa_\theta (\Pi_1\Eb)_r) \in V_2$ by construction (see Proposition 1). Then, 

$$
 \begin{alignat}{2}
 \int_0^{r^*_n} \int_0^{2\pi} (\pa_r (\Pi_1\Eb)_\theta - \pa_\theta (\Pi_1\Eb)_r)(r,\theta)\,\tn dr \tn d\theta    &=  \int_0^{2\pi} \left[ (\Pi_1\Eb)_\theta(r^*_n,\theta) - (\Pi_1\Eb)_\theta(0,\theta) \right] \tn d\theta && \nonumber
 \\[1mm]
 &\stackrel{\eqref{def:Pi1:2}}{=} \int_0^{2\pi} \left[ E_\theta(r^*_n,\theta) - E_\theta(0,\theta) \right] \tn d\theta\,, &&\qquad \tn{for} \ n = 1,2,3\,, \nonumber 
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\pa_r (\Pi_1\Eb)_\theta - \pa_\theta (\Pi_1\Eb)_r)\,\tn d r\tn d \theta 
 &= \int_{\theta_{j_2}}^{\theta_{j_2+1}} \left[ (\Pi_1\Eb)_\theta(r_{j_1+1}, \theta) - (\Pi_1\Eb)_\theta(r_{j_1}, \theta) \right] \tn d r\tn d \theta &&  \nonumber
 \\[1mm]
 &\qquad - \int_{r_{j_1}}^{r_{j_1+1}} \left[ (\Pi_1\Eb)_r(r,\theta_{j_2+1}) - (\Pi_1\Eb)_r(r,\theta_{j_2+1}) \right] \tn d r\tn d \theta &&  \label{diag:5}
 \\[1mm]
 & \stackrel{\eqref{def:Pi1:1},\eqref{def:Pi1:2}}{=}
 \int_{\theta_{j_2}}^{\theta_{j_2+1}} \left[ E_\theta(r_{j_1+1}, \theta) - E_\theta(r_{j_1}, \theta) \right] \tn d r\tn d \theta &&   \nonumber
 \\[1mm]
 &\qquad - \int_{r_{j_1}}^{r_{j_1+1}} \left[ E_r(r,\theta_{j_2+1}) - E_r(r,\theta_{j_2+1}) \right] \tn d r\tn d \theta
 \,, &&\qquad \tn{for} \ j_1 \geq 2\,.  \nonumber
 \end{alignat}
$$

On the other hand, 

$$
\begin{alignat}{2}
 \int_0^{r^*_n} \int_0^{2\pi} (\Pi_2 \curl\, \Eb)(r,\theta)\,\tn dr \tn d\theta &\stackrel{\eqref{def:Pi2}}{=} \int_0^{r^*_n} \int_0^{2\pi} \curl\, \Eb(r,\theta)\,\tn dr \tn d\theta
 = \int_0^{2\pi} \left[ E_\theta(r^*_n,\theta) - E_\theta(0,\theta) \right] \tn d\theta \,, && \qquad \tn{for} \ n = 1,2,3\,,  \nonumber
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_2 \curl\, \Eb)(r,\theta)\,\tn d r\tn d \theta &\stackrel{\eqref{def:Pi2}}{=}  \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}} \curl\, \Eb (r,\theta)\,\tn d r\tn d \theta &&  \label{diag:6}
 \\[1mm]
 &= \int_{\theta_{j_2}}^{\theta_{j_2+1}} \left[ E_\theta(r_{j_1+1}, \theta) - E_\theta(r_{j_1}, \theta) \right] \tn d r\tn d \theta &&   \nonumber
 \\[1mm]
 &\qquad - \int_{r_{j_1}}^{r_{j_1+1}} \left[ E_r(r,\theta_{j_2+1}) - E_r(r,\theta_{j_2+1}) \right] \tn d r\tn d \theta\,, && \qquad \tn{for} \ j_1 \geq 2\,. \nonumber
 \end{alignat}
$$

As \eqref{diag:5} is equal to \eqref{diag:6}, we have proved that the right part of the diagram commutes too, 

$$
\be
 \Pi_2 \curl = \curl\,\Pi_1\,.
\ee
$$

Therefore, the whole diagram commutes.
