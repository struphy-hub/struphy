# FEEC on domains with a polar singularity

## Preliminaries

The FEEC formalism runs into severe troubles when the mapping $F:U\to\Omega$ from the logical domain to the physical domain has a polar singularity. Let us consider the generic example of polar coordinates in 2D and let $(r,\theta) \in U$ denote the logical (polar) coordinates on the rectangle $U=[0,1]\times [0,2\pi)$, whereas $(x,y) \in \Omega$ stand for the physical coordinates on the unit disc $\Omega = S^1$:

$$
\be
 F: U\to\Omega\,,\qquad  (r,\theta) \mapsto (x,y)\,, \qquad \begin{pmatrix} x \\ y \end{pmatrix} = 
 \begin{pmatrix} r \cos\theta \\ r\sin\theta \end{pmatrix}\,.
\ee
$$

The Jacobian and its inverse are given by

$$
\be
 DF = \begin{pmatrix}
  \pder{x}{r} & \pder{x}{\theta}
  \\
  \pder{y}{r} & \pder{y}{\theta}
 \end{pmatrix} =
 \begin{pmatrix}
  \cos\theta & -r\sin\theta
  \\
  \sin\theta & r \cos\theta
 \end{pmatrix}\,, \qquad
 DF^{-1} = \frac{1}{r}\begin{pmatrix}
  r\cos\theta & r\sin\theta
  \\
  - \sin\theta & \cos\theta
 \end{pmatrix} \,,
\ee
$$

from which follow the metric tensor and related quantities

$$
\be
 G = DF^\top DF = \begin{pmatrix}
 1 & 0 \\ 0 & r^2
 \end{pmatrix}\,,\qquad g = \det G = r^2\,,\qquad G^{-1} = \begin{pmatrix}
 1 & 0 \\ 0 & \frac{1}{r^2}
 \end{pmatrix}\,.
\ee
$$

Moreover, the basic 1-forms $dr$ and $d\theta$, which are the lines of $(DF)^{-1}$, read

$$
\be
 dr = \begin{pmatrix} \cos\theta \\ \sin\theta  \end{pmatrix}\,,\qquad d\theta = \frac{1}{r} \begin{pmatrix}   -\sin\theta \\ \cos\theta  \end{pmatrix}\,.
\ee
$$

The whole FEEC formalism breaks down at $r = 0$ because $d\theta$ is not defined at that point. Ignoring this fact, we still can construct the following 2D deRham complexes on the logical domain $U$:

where

$$
\ba
 H(\curl) &:= \{ \Eb \in (L^2(U))^2: \ \curl\, \Eb \in L^2(U) \}\,,
 \\[2mm]
 H(\div) &:= \{ \Eb \in (L^2(U))^2: \ \div\, \Eb \in L^2(U) \}\,,
\ea
$$

and the involved operators are defined, for $\phi:U\to\RR$ and $\Eb = (E_r,E_\theta): U\to\RR^2$, as

$$
\ba
 & \grad\, \phi := \begin{pmatrix} \pa_r \phi \\ \pa_\theta \phi \end{pmatrix}\,,\qquad \curl\, \Eb := \pa_r E_\theta - \pa_\theta E_r\,,  \label{def:grad}
 \\[2mm]
 &\grad^\perp\, \phi := \begin{pmatrix} \pa_\theta \phi \\ -\pa_r \phi \end{pmatrix}\,,\qquad \div\, \Eb := \pa_r E_r + \pa_\theta E_\theta  \,.
\ea
$$

Clearly $\curl\, \grad = 0$ and $\div\,\grad^\perp = 0$ as necessary for a complex. Moreover, it is not hard to see that the adjoint of $\grad$ is $-\div$ and the adjoint of $\grad^\perp$ is $\curl$ (assuming the boundary terms vanish). We can write down the weak form of the Poisson equation in terms of differential forms using the left complex from above: find $(E^1,\phi^0) \in H\Lambda^1 \times H\Lambda^0$ such that

$$
\begin{alignat}{3}
&(E^1,F^1) + (\dd \phi^0,F^1) &&= 0\qquad &&\forall\ F^1 \in H\Lambda^1
 \\[1mm]
 &-(E^1, \dd \psi^0) &&= (*n^2, \psi^0) \qquad &&\forall\ \psi^0 \in H\Lambda^0\,.
\end{alignat}
$$

This is coordinate-independent; therefore, assuming that the above scalar products are finite in the physical domain, they are also finite in any other coordinate frame, also in the polar coordinate frame, where technically 1-forms $F^1$ cannot be defined at the pole $r=0$. On the logical domain the above system reads

$$
\be \label{Poisson:df:weak}
\left\{
\begin{aligned}
& \Eb + \nabla \phi = 0\qquad \tn{in}\ H(\curl,U)\,,
 \\[1mm]
 -\int_U &\Eb^\top G^{-1} \nabla \psi \sqrt{g} \,dr d\theta  = o\int_U n\, \psi\,dr d\theta \qquad \forall\ \psi \in H^1(U)\,,
\end{aligned}
\right.
\ee
$$

where $\nabla = \grad$ as defined in \eqref{def:grad} we can insert the polar metric to obtain

$$
\be \label{singint}
 \int_U \Eb^\top G^{-1} \nabla \psi \sqrt{g} \,dr d\theta
 =
 \int_U \Eb^\top \begin{pmatrix}
 r & 0 \\ 0 & \frac{1}{r}
 \end{pmatrix} \nabla \psi \,dr d\theta \,.
\ee
$$

That this integral is finite can be seen from the transformation law of 1-forms, namely

$$
\be
 E^1 = DF^{-\top} \Eb = \widetilde \Eb\,,
\ee
$$

where $\widetilde \Eb = (E_x,E_y)$ are the components of the 1-form in Cartesian (physical) coordinates. It follows that

$$
\ba
 \int_U \Eb^\top G^{-1} \nabla \psi \sqrt{g} \,dr d\theta
 &= \int_U \Eb^\top DF^{-1} DF^{-\top} \nabla \psi \sqrt{g} \,dr d\theta
 \\[1mm]
 &= \int_U \widetilde \Eb^\top \widetilde{ \nabla \psi} \,dx dy < \infty\,, \nonumber
\ea
$$

as we assume the integral to exist in physical space. The seemingly singular term with $1/r$ in \eqref{singint} is finite because the second component of $\nabla \psi$, namely $\pa_\theta \psi$, tends to zero at the pole sufficiently fast:

$$
\be
 \begin{pmatrix} \pa_r \psi \\ \pa_\theta\psi \end{pmatrix} = \nabla \psi =  DF^{\top} \widetilde{\nabla \psi} =
 \begin{pmatrix}
  \cos \theta\, \pa_x \widetilde \psi + \sin\theta\, \pa_y \widetilde\psi
  \\
  -r \sin \theta\, \pa_x \widetilde\psi + r \cos\theta\, \pa_y \widetilde\psi
 \end{pmatrix}\,.
\ee
$$

Therefore, on the logical domain we have, for any $\psi \in H^1(U)$, 

$$
\be \label{limgrad}
 \lim_{r\to 0} \nabla \psi = \begin{pmatrix}
  \cos \theta\, \pa_x \widetilde\psi + \sin\theta\, \pa_y \widetilde\psi
  \\
  0
 \end{pmatrix}\,.
\ee
$$

If \eqref{limgrad} was satisfied also on the discrete level, i.e. on the level of the basis functions of $V_0$, we could construct a deRham sequence for polar domains as follows:

* Construct a basis of $V_0$ such that each basis function satisfies \eqref{limgrad}.
* Construct a basis of $V_1$ via the exterior derivative $dV_0$ of the $V_0$-basis.
* The component of $d\theta$ in the $V_1$-basis will be zero at the pole due to \eqref{limgrad}, such that everything is well-defined.
* Construct the projectors $\Pi_0$ and $\Pi_1$ such that the diagram is commuting.

## $C^0$-continuous basis function at the pole 

The recipe for constructing $C^n$-continuous basis functions at the pole can be found in 

* D. Toshniwal et al., _Multi-degree smooth polar splines: A framework for geometric modeling and isogeometric analysis_, Comput. Methods Appl. Mech. Engrg. 316 (2017) 1005â€“1061. 

We follow their ideas but need to construct only $C^0$-continuity at the pole, which is straightforward. Suppose we have a B-spline basis of degree $k$ on the $r$-domain $[0, 1]$, denoted $(N_{i_1}^k(r))_{i_1=0}^{n_1-1}$, and a B-spline basis of degree $k$ on the $\theta$-domain $[0,2\pi)$, denoted $(N_{i_2}^k(\theta))_{i_2=0}^{n_2-1}$. We could also take different degrees $k_1$ and $k_2$ for $r$ and $\theta$ but choose the same degree in both directions for convenience. We demand the B-splines to be clamped at $r=0$, i.e.

$$
\be
 N_{i_1}^k(0) = 
 \begin{cases}
  1\,, & \ \tn{if} \ i_1 = 0\,,
  \\[1mm]
  0\,, & \ \tn{otherwise}\,.
 \end{cases}
\ee
$$

The standard tensor product basis of $V_0 \subset H^1(U)$ with multi-index $\ib = (i_1,i_2)$ would read $\Lambda_\ib^0 = N_{i_1}^k(r) N_{i_2}^k(\theta)$. Note that the degrees of freedom (DOFs) with index $i_1 = 0$ are mapped to the pole in the physical domain. Following Toshniwal et al., a basis that is $C^0$-continuous in the physical domain can be constructed by summing all the basis functions that correspond to DOFs mapped to the pole. Due to the partition-of-unity property of the B-splines we have

$$
\be
 \sum_{i_2=0}^{n_2-1} N_0^k(r) N_{i_2}^k(\theta) = N_0^k(r)\,,
\ee
$$

which we take as the new basis function, replacing the $n_2$ old basis functions at $i_1 = 0$. The new, appropriate basis for $V_0 \subset H^1(U)$ reads

$$
\be \label{basis:V0}
 \Lambda^0_\ib(r,\theta) := \begin{cases}
 N_0^k(r)\,, & \ \tn{if} \ i_1 = 0
 \\[1mm]
 N_{i_1}^k(r) N_{i_2}^k(\theta)\,, & \ \tn{otherwise}\,,
 \end{cases}
\ee
$$

where we mean $\tn{dim}\, V_0 = n_1 n_2 - n_2 + 1$ and the index $\ib$ is accordingly.
Taking the exterior derivative of these Basis functions leads to the following basis for $V_1 \subset H(\curl,U)$,

$$
\ba
 \tn dr:\qquad \Lambda^1_{\ib,1}(r,\theta) &:= \begin{cases}
 D_1^k(r)\,, & \ \tn{if} \ i_1 = 0
 \\[1mm]
 D_{i_1}^k(r) N_{i_2}^k(\theta)\,, & \ \tn{otherwise}\,,
 \end{cases}  \label{basis:V1:r}
 \\[3mm]
 \tn d \theta:\qquad \Lambda^1_{\ib,2}(r,\theta) &:= \begin{cases}
 0\,, & \ \tn{if} \ i_1 = 0
 \\[1mm]
 N_{i_1}^k(r) D_{i_2}^k(\theta)\,, & \ \tn{otherwise}\,.
 \end{cases}  \label{basis:V1:theta}
\ea
$$

In particular, we remark that 

$$
\be
 \lim_{r\to 0} \Lambda^1_{\ib,2}(r,\theta) = 0 \qquad \forall \ \ib\,,
\ee
$$

such that the important property \eqref{limgrad} is satisfied by all basis functions of $V_1$. Finally, by taking the exterior derivative of the basis functions \eqref{basis:V1:r} and \eqref{basis:V1:theta}, which corresponds to taking the $\curl$, we obtain the basis for $V_2 \subset L^2(U)$:

$$
\be \label{basis:V2}
 \Lambda^2_\ib(r,\theta) := \begin{cases}
 0\,, & \ \tn{if} \ i_1 = 0
 \\[1mm]
 D_{i_1}^k(r) D_{i_2}^k(\theta)\,, & \ \tn{otherwise}\,.
 \end{cases}
\ee
$$

__Proposition 1. (Exact sequence property.)__ The 2D-$H(\curl)$-diagram satisfies the exact sequence property

$$
\be
 \Im\, \grad V_0 = \tn{Ker}\, \curl V_1
\ee
$$

if we choose as basis functions for $V_0$ and $V_1$ the functions defined in \eqref{basis:V0} and \eqref{basis:V1:r}-\eqref{basis:V1:theta}, respectively.

_Proof:_ We need to check this for each of the basis functions in \eqref{basis:V0}. For $i_1>0$ we have the standard case,

$$
\be
\grad\left[ N_{i_1}^k(r) N_{i_2}^k(\theta) \right] = \begin{pmatrix}
\left[ D_{i_1}^k(r) - D_{i_1+1}^k(r) \right] N_{i_2}^k(\theta)
\\
N_{i_1}^k(r) \left[ D_{i_2}^k(\theta) - D_{i_2+1}^k(\theta) \right]
\end{pmatrix} \,,
\ee
$$

and thus

$$
\be
 \curl\,\grad \left[ N_{i_1}^k(r) N_{i_2}^k(\theta) \right] = 
\left[ D_{i_1}^k(r) - D_{i_1+1}^k(r) \right] \left[ D_{i_2}^k(\theta) - D_{i_2+1}^k(\theta) \right]
 - \left[ D_{i_1}^k(r) - D_{i_1+1}^k(r) \right] \left[ D_{i_2}^k(\theta) - D_{i_2+1}^k(\theta) \right] = 0\,.
\ee
$$

For $i_1 = 0$ we have

$$
\be
\grad\, N_0^k(r)  = \begin{pmatrix}
 - D_1^k(r) 
\\
0
\end{pmatrix} \qquad \Rightarrow \qquad \curl\,\grad\,  N_0^k(r) = 0\,,
\ee
$$

which completes the proof. 

## Projectors

We now want to ensure the commuting property of the diagram with the new basis, which is not a tensor product basis anymore. For $a \in H^1$ we have $(\Pi_0 a) \in V_0$, expressed in the basis \eqref{basis:V0} as

$$
\be
 (\Pi_0 a)(r,\theta) = a_0\, N_0^k(r) + \sum_{i_1 =1}^{n_1-1}\sum_{i_2 = 0}^{n_2-1} a_\ib\, N_{i_1}^k(r) N_{i_2}^k(\theta)\,.
\ee
$$

We define the coefficients $(a_0,a_\ib)$ by interpolation, but take the $\theta$-average of interpolations at the pole $r_{j_1 = 0}=0$:

$$
\be \label{def:Pi0}
 \begin{cases}
 (\Pi_0 a)(0,\theta) = a_0 = \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} a(0,\theta_{j_2})\,, & \ \tn{for} \ j_1 = 0\,,
 \\[2mm]
 (\Pi_0 a)(r_{j_1}, \theta_{j_2}) = a(r_{j_1}, \theta_{j_2})\,, & \ \tn{otherwise}\,.
 \end{cases}
\ee
$$

On the other hand, for $\Eb=(E_1,E_2) \in H(\curl,U)$ we have $(\Pi_1 \Eb) \in V_1$, expressed in the basis \eqref{basis:V1:r}-\eqref{basis:V1:theta} as

$$
\be
 (\Pi_1 \Eb)(r,\theta) = 
 \begin{pmatrix}
 E_{0,1}\, D_1^k(r) + \sum_{i_1 =1}^{n_1-1}\sum_{i_2 = 0}^{n_2-1} E_{\ib,1}\, D_{i_1}^k(r) N_{i_2}^k(\theta)
 \\[1mm]
 \sum_{i_1 =1}^{n_1-1}\sum_{i_2 = 0}^{n_2-1} E_{\ib,2}\, N_{i_1}^k(r) D_{i_2}^k(\theta)
 \end{pmatrix} \,.
\ee
$$

We define the coefficients $(E_{0,1}, E_{\ib,1})$ of the first component $(\Pi_1 \Eb)_1$ of $\Pi_1 \Eb$ via

$$
\be \label{def:Pi1:1}
 \begin{cases}
 \int_0^{r_1}(\Pi_1 \Eb)_1(r,\theta_{j_2})\,\tn d r = \mathring{E_1}(r_1, \theta_{j_2}) - \frac{1}{n_2} \sum_{j_2 = 0}^{n_2-1} \mathring{E_1}(0, \theta_{j_2}) \,, & \ \tn{if} \ j_1 = 0\,, 
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}}(\Pi_1 \Eb)_1(r,\theta_{j_2})\,\tn d r = \int_{r_{j_1}}^{r_{j_1+1}}E_1(r,\theta_{j_2})\,\tn d r \,, & \ \tn{otherwise}\,,
 \end{cases}
\ee
$$

where $\mathring{E_1}$ denotes the $r$-primitive of the first component $E_1$ of $\Eb$, 

$$
\be
 \mathring{E_1}(r,\theta) := \int_0^r E_1(s,\theta)\,\tn d s - \int_0^1 E_1(s,\theta)\,\tn d s\,.
\ee
$$

Remark that the constant of integration in this primitive integral has been set to the $r$-average over the domain $[0,1]$. This will be sufficient to achieve the commuting property. The coefficients $(E_{\ib,2})$ of the second component $(\Pi_1 \Eb)_2$ of $\Pi_1 \Eb$ are defined by 

$$
\be \label{def:Pi1:2}
 \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_1 \Eb)_2(r_{j_1},\theta)\,\tn d \theta = \int_{\theta_{j_2}}^{\theta_{j_2+1}}E_2(r_{j_1},\theta)\,\tn d \theta \,, \quad \tn{for}\ \ j_1 >0 \,.
\ee
$$

Finally, for $b \in L^2$ we have $(\Pi_2 b) \in V_2$, expressed in the basis \eqref{basis:V2} as

$$
\be
 (\Pi_2 b)(r,\theta) = \sum_{i_1 =1}^{n_1-1}\sum_{i_2 = 0}^{n_2-1} b_\ib\, D_{i_1}^k(r) D_{i_2}^k(\theta)\,.
\ee
$$

We define the coefficients $(b_\ib)$ by histopolation,

$$
\be \label{def:Pi2}
 \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_2 b)(r,\theta)\,\tn d r\tn d \theta =    \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}} b(r,\theta)\,\tn d r\tn d \theta\,,\qquad \tn{for} \ j_1 >1\,. 
\ee
$$


__Proposition 2. (Commuting diagram for polar domain.)__ The 2D-$H(\curl)$-diagram commutes if

* we choose as basis functions for $V_0$, $V_1$ and $V_2$ the functions defined in \eqref{basis:V0}, \eqref{basis:V1:r}-\eqref{basis:V1:theta} and \eqref{basis:V2}, respectively, 
* we define the projector $\Pi_0$ as in \eqref{def:Pi0}, the projector $\Pi_1$ as in \eqref{def:Pi1:1}-\eqref{def:Pi1:2} and the projector $\Pi_2$ as in \eqref{def:Pi2}, 
* we impose Dirichlet boundary conditions independent of $\theta$ at the outer boundary $r=1$, i.e. for $\psi \in H^1(U)$ we demand

$$
\be \label{bc:Dirichlet}
 \psi(1,\theta) = c \in \RR\,.
\ee
$$

_Proof:_ For $\psi \in H^1$ we have $(\Pi_0\psi) \in V_0$ and $\grad (\Pi_0\psi) = (\pa_r (\Pi_0\psi),\, \pa_\theta (\Pi_0\psi))^\top \in V_1$ by construction (see Proposition 1). Hence, for the first component of the gradient we have

$$
\be \label{prf1}
\int_{r_{j_1}}^{r_{j_1+1}} \pa_r (\Pi_0\psi) (r,\theta_{j_2})\,\tn d r = (\Pi_0\psi) (r_{j_1+1},\theta_{j_2}) - (\Pi_0\psi) (r_{j_1},\theta_{j_2}) =
 \begin{cases}
   \psi(r_1,\theta_{j_2}) - \frac{1}{n_2} \sum_{i_2 = 0}^{n_2-1} \psi(0,\theta_{j_2}) \,, & \ \tn{if} \ j_1 = 0\,, 
 \\[2mm]
   \psi(r_{j_1+1},\theta_{j_2}) - \psi(r_{j_1},\theta_{j_2}) \,, & \ \tn{otherwise}\,,
 \end{cases}
\ee
$$

and for the second component of the gradient we have

$$
\be \label{prf2}
 \int_{\theta_{j_2}}^{\theta_{j_2+1}}\pa_\theta (\Pi_0\psi)(r_{j_1},\theta)\,\tn d \theta = (\Pi_0\psi)(r_{j_1},\theta_{j_2+1}) - (\Pi_0\psi)(r_{j_1},\theta_{j_2}) = \begin{cases}
   0\,, & \ \tn{if} \ j_1 = 0\,, 
 \\[2mm]
   \psi(r_{j_1}, \theta_{j_2+1}) - \psi(r_{j_1}, \theta_{j_2})\,, & \ \tn{otherwise}\,,
 \end{cases}
\ee
$$

where we used the definition \eqref{def:Pi0} of the projector $\Pi_0$ to get the second equality for both components. On the other hand we have $\grad\,\psi \in H(\curl)$ and the projection $(\Pi_1\grad\,\psi) \in V_1$, which is defined component-wise by \eqref{def:Pi1:1} and \eqref{def:Pi1:2}, respetively. For the first component we compute the primitive

$$
\ba
 \mathring{(\pa_r\psi)}(r,\theta) &= \int_0^r \pa_r\psi(s,\theta)\,\tn d s - \int_0^1 \pa_r\psi(s,\theta)\,\tn d s
 \\[1mm]
 &= \psi(r,\theta) - \psi(0,\theta) - \psi(1,\theta) + \psi(0,\theta) 
 \\[3mm]
 &= \psi(r,\theta) - c\,,
\ea
$$

where we used the Dirichlet boundary condition \eqref{bc:Dirichlet}. This leads for the first component to

$$
\be \label{prf3}
 \begin{cases}
 \int_0^{r_1}(\Pi_1\grad\,\psi)_1(r,\theta_{j_2})\,\tn d r = \mathring{(\pa_r \psi)}(r_1, \theta_{j_2}) - \frac{1}{n_2} \sum_{i_2 = 0}^{n_2-1} \mathring{(\pa_r \psi)}(0, \theta_{j_2}) = \psi(r_1,\theta) - \frac{1}{n_2} \sum_{i_2 = 0}^{n_2-1} \psi(0, \theta_{j_2})  \,, & \ \tn{if} \ j_1 = 0\,, 
 \\[3mm]
 \int_{r_{j_1}}^{r_{j_1+1}}(\Pi_1\grad\,\psi)_1(r,\theta_{j_2})\,\tn d r = \int_{r_{j_1}}^{r_{j_1+1}} \pa_r \psi(r,\theta_{j_2})\,\tn d r = \psi(r_{j_1+1},\theta_{j_2}) - \psi(r_{j_1},\theta_{j_2}) \,, & \ \tn{otherwise}\,,
 \end{cases}
\ee
$$

and for the second component we obtain

$$
\be \label{prf4}
 \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_1 \grad\,\psi)_2(r_{j_1},\theta)\,\tn d \theta = \int_{\theta_{j_2}}^{\theta_{j_2+1}}\pa_\theta\psi(r_{j_1},\theta)\,\tn d \theta = \psi(r_{j_1},\theta_{j_2+1}) - \psi(r_{j_1},\theta_{j_2})\,,\qquad \tn{for} \ j_1 >0\,. 
\ee
$$

We note that \eqref{prf1} is the same as \eqref{prf3} and that \eqref{prf2} is the same as \eqref{prf4}. As these relations uniquely define an element in $V_1$, we proved that the left part of the diagram commutes. Now let us prove the right part. For $\Eb \in H(\curl)$ we have $(\Pi_1\Eb) \in V_1$ and $\curl\, (\Pi_1\Eb) = \left[\pa_r (\Pi_1\Eb)_2 - \pa_\theta (\Pi_1\Eb)_1 \right] \in V_2$ by construction (see Proposition 1). Hence,

$$
\ba
 \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}}\curl\, (\Pi_1\Eb)(r,\theta)\,\tn d r\tn d \theta &=   \int_{\theta_{j_2}}^{\theta_{j_2+1}}\left[(\Pi_1\Eb)_2(r_{j_1+1},\theta) - (\Pi_1\Eb)_2(r_{j_1},\theta) \right]\,\tn d\theta \label{prf5}
 \\[1mm]
 &\quad - \int_{r_{j_1}}^{r_{j_1+1}}\left[ (\Pi_1\Eb)_1(r,\theta_{j_2+1}) - (\Pi_1\Eb)_1(r,\theta_{j_2}) \right]\,\tn d r  \nonumber
 \\[1mm]
 &= \int_{\theta_{j_2}}^{\theta_{j_2+1}}\left[E_2(r_{j_1+1},\theta) - E_2(r_{j_1},\theta) \right]\,\tn d\theta \nonumber
 \\[1mm]
 &\quad - \int_{r_{j_1}}^{r_{j_1+1}}\left[ E_1(r,\theta_{j_2+1}) - E_1(r,\theta_{j_2}) \right]\,\tn d r\,,\qquad \tn{for} \ j_1 >1\,.  \nonumber
\ea
$$

Here, we used the definition of the projector $\Pi_1$ to get to the two last lines. On the other hand we have $\curl\,\Eb \in L^2$ and the projection $(\Pi_2\curl\,\Eb) \in V_2$, defined by

$$
\ba
 \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}}(\Pi_2\curl\,\Eb)(r,\theta)\,\tn d r\tn d \theta &= \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}} \curl\,\Eb (r,\theta)\,\tn d r\tn d \theta   \label{prf6}
 \\[1mm]
 &= \int_{r_{j_1}}^{r_{j_1+1}} \int_{\theta_{j_2}}^{\theta_{j_2+1}} \left[\pa_r E_2(r,\theta) - \pa_\theta E_1(r,\theta) \right]\,\tn d r\tn d \theta  \nonumber  
 \\[1mm]
 &= \int_{\theta_{j_2}}^{\theta_{j_2+1}}\left[E_2(r_{j_1+1},\theta) - E_2(r_{j_1},\theta) \right]\,\tn d\theta \nonumber
 \\[1mm]
 &\quad - \int_{r_{j_1}}^{r_{j_1+1}}\left[ E_1(r,\theta_{j_2+1}) - E_1(r,\theta_{j_2}) \right]\,\tn d r\,,\qquad \tn{for} \ j_1 >1\,.  \nonumber
\ea
$$

We note that \eqref{prf5} is the same as \eqref{prf6}. As these relations uniquely define an element in $V_2$, we proved that the right part of the diagram commutes, and therefore the whole diagram commutes.
