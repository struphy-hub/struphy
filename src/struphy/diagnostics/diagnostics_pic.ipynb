```python
import os
import struphy
import numpy as np
from matplotlib import pyplot as plt

path_out = os.path.join(struphy.__path__[0], "io/out", "sim_1")

print(path_out)
os.listdir(path_out)
```

```python
data_path = os.path.join(path_out, "post_processing")

os.listdir(data_path)
```

```python
t_grid = np.load(os.path.join(data_path, "t_grid.npy"))
t_grid
```

```python
f_path = os.path.join(data_path, "kinetic_data", "ions", "distribution_function")

print(os.listdir(f_path))
```

```python
path = os.path.join(f_path, "e1")
print(os.listdir(path))
```

```python
grid = np.load(os.path.join(f_path, "e1/", "grid_e1.npy"))
f_binned = np.load(os.path.join(f_path, "e1/", "f_binned.npy"))
delta_f_e1_binned = np.load(os.path.join(f_path, "e1/", "delta_f_binned.npy"))

print(grid.shape)
print(f_binned.shape)
print(delta_f_e1_binned.shape)
```

```python
steps = list(np.arange(10))

plt.figure(figsize=(12, 5 * len(steps)))
for n, step in enumerate(steps):
    plt.subplot(len(steps), 2, 2 * n + 1)
    plt.plot(grid, f_binned[step], label=f"time = {t_grid[step]}")
    plt.xlabel("e1")
    # plt.ylim([.5, 1.5])
    plt.title("full-f")
    plt.subplot(len(steps), 2, 2 * n + 2)
    plt.plot(grid, delta_f_e1_binned[step], label=f"time = {t_grid[step]}")
    plt.xlabel("e1")
    # plt.ylim([-3e-3, 3e-3])
    plt.title(r"$\delta f$")
    plt.legend()
```

```python
path = os.path.join(f_path, "e1_v1")
print(os.listdir(path))
```

```python
grid_e1 = np.load(os.path.join(f_path, "e1_v1/", "grid_e1.npy"))
grid_v1 = np.load(os.path.join(f_path, "e1_v1/", "grid_v1.npy"))
f_binned = np.load(os.path.join(f_path, "e1_v1/", "f_binned.npy"))
delta_f_binned = np.load(os.path.join(f_path, "e1_v1/", "delta_f_binned.npy"))

print(grid_e1.shape)
print(grid_v1.shape)
print(f_binned.shape)
print(delta_f_binned.shape)
```

```python
steps = list(np.arange(10))

plt.figure(figsize=(12, 5 * len(steps)))
for n, step in enumerate(steps):
    plt.subplot(len(steps), 2, 2 * n + 1)
    plt.pcolor(grid_e1, grid_v1, f_binned[step].T, label=f"time = {t_grid[step]}")
    plt.xlabel("$e1$")
    plt.ylabel(r"$v_\parallel$")
    plt.title("full-f")
    plt.legend()
    plt.colorbar()
    plt.subplot(len(steps), 2, 2 * n + 2)
    plt.pcolor(grid_e1, grid_v1, delta_f_binned[step].T, label=f"time = {t_grid[step]}")
    plt.xlabel("$e1$")
    plt.ylabel(r"$v_\parallel$")
    plt.title(r"$\delta f$")
    plt.legend()
    plt.colorbar()
```

```python
fields_path = os.path.join(data_path, "fields_data")

print(os.listdir(fields_path))
```

```python
import pickle

with open(os.path.join(fields_path, "grids_phy.bin"), "rb") as file:
    x_grid, y_grid, z_grid = pickle.load(file)

print(type(x_grid))
print(x_grid.shape)
```

```python
with open(os.path.join(fields_path, "em_fields", "phi_phy.bin"), "rb") as file:
    phi = pickle.load(file)

plt.figure(figsize=(12, 12))

steps = [0, 20, 40, -1]
for n, step in enumerate(steps):
    t = t_grid[step]
    print(phi[t][0].shape)
    plt.subplot(2, 2, n + 1)
    plt.plot(x_grid[:, 0, 0], phi[t][0][:, 0, 0], label=f"time = {t}")
    plt.xlabel("x")
    plt.ylabel(r"$\phi$(x)")
    plt.legend()
```
