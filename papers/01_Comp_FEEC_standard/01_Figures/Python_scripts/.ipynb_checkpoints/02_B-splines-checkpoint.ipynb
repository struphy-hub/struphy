```python
import numpy as np
import matplotlib.pyplot as plt
import Utilitis_HybridCode as utils
from copy import deepcopy

plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')

Lz = 1.
p = 1
Nel = 5

bsp, Nbase, quad_points, weights = utils.createBasis(Lz, Nel, p, bcs = 1)
el_b = np.linspace(0, Lz, Nel + 1) 

zplot = np.linspace(0, Lz, 200)

f1 = plt.figure()
f1.set_figheight(2)
f1.set_figwidth(3)

colors = []

for i in range(Nbase - p):
    pl = plt.plot(zplot, bsp(zplot, i), linewidth = 1)
    colors.append(pl[0].get_color())
    
    if i <= Nel - 2 and i > 0:
        plt.plot(el_b[i]*np.ones(10), np.linspace(0, 0.8, 10), 'k--')

color_counter = 0
for i in range(Nel, Nbase):
    plt.plot(zplot, bsp(zplot, i), linewidth = 1, color = colors[color_counter])
    color_counter += 1
        

plt.plot(el_b[-2]*np.ones(10), np.linspace(0, 0.8, 10), 'k--', label = 'element boundaries')
plt.plot(quad_points, np.zeros(len(quad_points)), 'ro', label = 'quadrature points', markersize = 3)


plt.xlabel('$z$')
plt.ylim((-0.25, 1.75))
plt.xlim((0, 1))
plt.legend()
#plt.savefig('/home/florian/Schreibtisch/Masterarbeit/08_Paper/01_Figures/Bsplines.pgf', bbox_inches = 'tight')
```
