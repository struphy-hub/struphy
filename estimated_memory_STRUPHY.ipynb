```python
import numpy as np
```

```python
N_vec = np.array([2**n for n in range(3,10)])
p_vec = [n+1 for n in range(8)]
ppcell = 1000

print('{0:4s} {1:6s} {2:10s} {3:11s} {4:11s} {5:11s}'.format('   p','     N','   N**3','  mem_MHD (GB)',' mem_PIC (GB)',' mem_tot (GB)'))
for p in p_vec:
    print()
    for N in N_vec:
        mem_mass0 = N**3*(2*p+1)*64/8*1e-9
        mem_grad_line = N**3*(2)*64/8*1e-9
        mem_MHD = mem_mass0*59 + mem_grad_line*12
        mem_PIC = 7*ppcell*N**3*64/8*1e-9
        print('{0:4d} {1:6d}   {2:10d}   {3:10.4f}   {4:10.4f}   {5:10.4f}'.format(
            p, N, N**3, mem_MHD, mem_PIC, mem_MHD+mem_PIC))
```
